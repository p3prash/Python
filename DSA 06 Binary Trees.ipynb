```python
class BinaryTreeNode :
  def __init__(self, data) :
    self.data = data
    self.left = None
    self.right = None
```

```python
def printTree(root) :
  if root == None :
    return
  print(root.data)
  printTree(root.left)
  printTree(root.right)
```

```python
#Taking level-order input using fast I/O method
import queue
def takeInput():
    levelOrder = list(map(int, input().strip().split(" ")))
    start = 0
    length = len(levelOrder)
    root = BinaryTreeNode(levelOrder[start])
    start += 1

    q = queue.Queue()
    q.put(root)

    while not q.empty():
        currentNode = q.get()

        leftChild = levelOrder[start]
        start += 1

        if leftChild != -1:
            leftNode = BinaryTreeNode(leftChild)
            currentNode.left =leftNode
            q.put(leftNode)

        rightChild = levelOrder[start]
        start += 1

        if rightChild != -1:
            rightNode = BinaryTreeNode(rightChild)
            currentNode.right =rightNode
            q.put(rightNode)

    return root
X = takeInput()
PrintTreeDetailed(X)
```

```python
def isNodePresent(root, x):
    # Write your code here.
    if root == None :
        return 
    if root.data == x :
        return True
    
    isLeft = isNodePresent(root.left, x)
    if isLeft :
        return True
    isRight = isNodePresent(root.right, x)
    if isRight :
        return True
    else :
      return False

# Main
root = takeInput()
x = int(input())
root = isNodePresent(root, x)
print(root)
```

```python
def Print_Level_Wise(root):
    if root==None:
        return
    inputQ = queue.Queue()
    outputQ = queue.Queue()
    inputQ.put(root)

    while not inputQ.empty():    
        while not inputQ.empty():   
            curr = inputQ.get()
            print(curr.data, end=':')
            if curr.left == None :
              print('L:-1',end=',',sep='')
            else :  
              print('L:' ,curr.left.data,end=',',sep='')

            if curr.right == None :
              print('R:-1',sep='')
            else :  
              print('R:' ,curr.right.data,sep='')

            if curr.left!=None:
                outputQ.put(curr.left)
            if curr.right!=None:
                outputQ.put(curr.right)
       
        inputQ, outputQ = outputQ, inputQ
        
root = takeInput()
Print_Level_Wise(root)
```

```python
def printLevelWise(root):
    if root==None:
        return
    inputQ = queue.Queue()
    outputQ = queue.Queue()
    inputQ.put(root)

    while not inputQ.empty():    
        while not inputQ.empty() :
            curr = inputQ.get()
            print(curr.data,end=" ")

            if curr.left!=None:
                outputQ.put(curr.left)
            if curr.right!=None:
                outputQ.put(curr.right)
        
        inputQ, outputQ = outputQ, inputQ
        print()
        
root = takeInput()
printLevelWise(root)
```

```python
def isTreeBalanced(root) :
  if root == None :
    return True
  LH = height(root.left)
  RH = height(root.right)
  if LH-RH>1 or RH-LH>1 :
    return False
  isLbalanced = isTreeBalanced(root.left)
  isRbalanced = isTreeBalanced(root.right)
  if isLbalanced and isRbalanced :
    return True
  else :
    return False

root = takeInput()
printLevelWise(root)
isTreeBalanced(root)
```

```python
def DiameterOfBTree(root) :
  if root == None :
    return 0
  LH = height(root.left)
  RH = height(root.right)
  Dia = 1+LH+RH
  LDia = DiameterOfBTree(root.left)
  RDia = DiameterOfBTree(root.right)
  if Dia > LDia and Dia > RDia :
    return Dia
  elif LDia > Dia and LDia > RDia :
    return LDia
  else :
    return RDia

root = takeInput()
Dia = DiameterOfBTree(root)
print(Dia)
```

```python
def longestPath(root):
    if (root == None):
        return []
 
    # Recursive call on root.right
    rightvect = longestPath(root.right)
 
    # Recursive call on root.left
    leftvect = longestPath(root.left)

    if (len(leftvect) > len(rightvect)):
        leftvect.append(root.data)
    else:
        rightvect.append(root.data)

    if len(leftvect) > len(rightvect):
        return leftvect
 
    return rightvect
```
