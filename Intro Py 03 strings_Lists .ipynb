```python
# print column wise sum of a given 2d list
def print_column_wise_Sum(A):
  m = len(A)
  if m == 0:
    return
  n = len(A[0])
  for j in range(n):
    Sum_j = 0
    for i in range(m):
      Sum_j = Sum_j + A[i][j]  
    print(Sum_j)      
```

```python
def removeConsecutiveDuplicates(string):    
    n = len(string)

    if (n < 2) :
        return
    j = 0
# Traversing string
    for i in range(n): 
        if (string[j] != string[i]):
            j += 1
            string[j] = string[i] 
    j += 1
    string = string[:j]
    return string
     
# Driver Code
if __name__ == '__main__':       
    string = input()
    string = list(string.rstrip())
    string = removeConsecutiveDuplicates(string)
    print(*string,sep='')
```

```python
def MaxSum(X, Y):
    i,j = 0, 0
    m, n = len(X), len(Y)
    maxsum, S1, S2 = 0, 0, 0

    while (i < m and j < n):
        if (X[i] < Y[j]):
            S1 = S1 + X[i]
            i+=1
            
        elif (X[i] > Y[j]):
            S2 = S2 + Y[j]
            j+=1
            
        else:
            maxsum += max(S1,S2) + X[i]
            S1=0; S2=0
            i+=1; j+=1

    while i < m:
        S1 += X[i]
        i += 1

    while j < n:
        S2 += Y[j]
        j += 1
 
    maxsum += max(S1, S2)
    return maxsum     

m = int(input())
X = list(map(int,input().strip().split()))[:m]
n = int(input())
Y = list(map(int,input().strip().split()))[:n]
print(MaxSum(X, Y))
```

```python
def isPermutation(str1, str2) :
  n1 = len(str1)
  n2 = len(str2)
   
  if (n1 != n2):
    return False
  frequency = [0] * 256
  
  for i in range(n1):
    ch = ord(str1[i])
    frequency[ch] += 1

  for j in range(n2):
    ch = ord(str2[j])
    frequency[ch] -= 1

  for k in range(256):
    if frequency[k] != 0:
      return False    
  return True
#main
str1 = input()
str2 = input()
ans = isPermutation(str1, str2)
if ans :
    print('true')
else :
    print('false')
```

```python
def multiply(a, b):

    mod = int(1e9 + 7)
    c = [[0 for i in range(2)] for j in range(2)]
    for i in range(2):
        for j in range(2):
            for k in range(2):
                c[i][k] = (c[i][k] + a[i][j] * b[j][k]) % mod

    return c
# Function to calclate n'th power of a matrix.
def matrix_power(coef, n):
  
    res = [[1, 0], [0, 1]]
    # While loop till n > 0.
    while n > 0:
        
        if (n % 2 != 0):
            res = multiply(res, coef)    
        # Multiply coef with coef and update n to n//2.
        coef = multiply(coef, coef)
        n //= 2
        
    return res

def fibonacciNumber(n):
    
    coef = [[0, 1], [1, 1]]
    # Calculating the (n-1)th power of the coef matrix.
    res = matrix_power(coef, n - 1)
    # Return the bottom right element of the resultant matrix.
    return res[1][1]

n = int(input())
print(fibonacciNumber(n))
```

```python
def fibonacciNumber(n):
    mod = int(1e9 + 7)
    # Create an array to store solution
    dp=[0 for i in range(n + 1)]
    dp[1] = 1
    # Iterate i from 2 to N and find ith Fibonacci Number 
    for i in range(2, n + 1):
        dp[i] = (dp[i - 1] + dp[i - 2]) % mod
    # Return the Nth Fibonacci Number
    return dp[n]
n = int(input()) 
print(fibonacciNumber(n)) 
```
