```python
import random
def start_game():
    mat = []
    for i in range(4):
        mat.append([0]*4)
    return mat

def Add_New2(mat) :
    r = random.randint(0,3)
    c = random.randint(0,3)
    while mat[r][c] != 0 :
        r = random.randint(0,3)
        c = random.randint(0,3)
    mat[r][c] = 2

def Reverse(mat) :
    new_Mat = []
    for i in range(4) :
        new_Mat.append([])
        for j in range(4) :
            new_Mat[i].append(mat[i][4-j-1])
    return new_Mat

def Transpose(mat) :
    new_Mat = []
    for i in range(4) :
        new_Mat.append([])
        for j in range(4) :
            new_Mat[i].append(mat[j][i])
    return new_Mat

def Merge(mat) :
    CH = False
    for i in range(4) :
        for j in range(3) :
            if mat[i][j] == mat[i][j+1] and mat[i][j] != 0 :
                mat[i][j] = mat[i][j] * 2
                mat[i][j+1] = 0
                CH = True
    return mat, CH

def Compress(mat) :
    N_mat = []
    CH = False
    for i in range(4) :
        N_mat.append([0]*4) 
    
    for i in range(4) :
        pos = 0
        for j in range(4) :
            if mat[i][j] != 0 :
                N_mat[i][pos] = mat[i][j]
                if j != pos :
                    CH = True
                pos+=1
    return N_mat, CH

def move_left(grid):
    C_grid, CH1 = Compress(grid)
    M_grid, CH2 = Merge(C_grid)
    CH = CH1 or CH2 
    F_grid = Compress(M_grid)
    return F_grid, CH

def move_right(grid):
    R_grid = Reverse(grid)
    C_grid, CH1 = Compress(R_grid)
    M_grid, CH2 = Merge(C_grid)
    CH = CH1 or CH2
    RC_grid = Compress(M_grid)
    F_grid = Reverse(RC_grid)
    return F_grid, CH

def move_up(grid):
    T_grid = Transpose(grid)
    C_grid, CH1 = Compress(T_grid)
    M_grid, CH2 = Merge(C_grid)
    CH = CH1 or CH2
    RC_grid = Compress(M_grid)
    F_grid = Transpose(RC_grid)
    return F_grid, CH

def move_down(grid) :
    T_grid = Transpose(grid)
    R_grid = Reverse(T_grid)
    C_grid, CH1 = Compress(R_grid)
    M_grid, CH2 = Merge(C_grid)
    RC_grid = Compress(M_grid)
    RR_grid = Reverse(RC_grid)
    F_grid = Transpose(RR_grid)
    return F_grid, CH

mat = start_game()
mat[1][3] = 2
mat[2][2] = 2
mat[3][0] = 4
mat[3][1] = 8
mat[2][1] = 4
inputs = [int(ele) for ele in input().split()]
for ele in inputs:
    if ele == 1:
        mat = move_up(mat)
    elif ele == 2:
        mat = move_down(mat)
    elif ele == 3:
        mat = move_left(mat)
    else:
        mat = move_right(mat)
    print(mat)
```

```python
def get_Curr_state(mat) :
  # winning condition
    for i in range(4) :
        for j in range(4) :
            if (mat[i][j] == 2048) :
                return 'WON'

  # game still on
    for i in range(4) :
        for j in range(4) :
            if (mat[i][j] == 0) :
                return 'GAME NOT OVER'
    for i in range(3) :
        for j in range(3) :
            if mat[i][j] == mat[i+1][j] or mat[i][j] == mat[i][j+1] :
                return 'GAME NOT OVER'
    for j in range(3) :
        if mat[3][j] == mat[3][j+1] :
            return 'GAME NOT OVER'
    for i in range(3) :
        if mat[i][3] == mat[i+1][3] :
            return 'GAME NOT OVER'
    
  # game over
    return 'LOST'
```
