```python
def Subsequences(string) :
    out = []
    if len(string) == 0 :
        out.append("")
        return out
    SubStr = string[1:]
    Smallout = Subsequences(SubStr)

    for sub in Smallout :
        out.append(sub)
    for sub in Smallout :
        Subextra = string[0] + sub
        out.append(Subextra)
    return out

string = input()
ans = Subsequences(string)
for i in ans :
    print(i, end=" ")
```

```python
def getString(d) :
    if d==2 :
      return "abc"
    if d==3 :
      return "def"
    if d==4 :
      return "ghi"
    if d==5 :
      return "jkl"
    if d==6 :
      return "mno"
    if d==7 :
      return "pqrs"
    if d==8 :
      return "tuv"
    if d==9 :
      return "wxyz"
    return " "

def Keypad(num) :
    out = []
    if num == 0 :
        out.append("")
        return out
    smallNum = num // 10
    last = num % 10
    smallout = Keypad(smallNum)
    text = getString(last)

    for s in smallout :
        for char in text :
            opt = s + char
            out.append(opt)
    return out

num = int(input())
print(Keypad(num))
```

```python
def PrintFact(n, ans=1) :
    if n == 0 :
        print(ans)
        return
    ans = ans * n
    PrintFact(n-1, ans)
n = int(input())
PrintFact(n)
```

```python
import sys
def Printmin(L, Min = sys.maxsize) :
    if len(L) == 0 :
        print(Min)
        return
    Min = min(Min, L[0])
    Printmin(L[1:], Min)
L = [int(i) for i in input().split()]
Printmin(L)
```

```python
def PrintSubsequence(Str, out="") :
    if len(Str) == 0 :
        print(out, end=" ")
        return
    PrintSubsequence(Str[1:], out)
    out = out + Str[0]
    PrintSubsequence(Str[1:], out)
Str = input()
PrintSubsequence(Str)
```

```python
def PrintKeypad(Digit, out = "") :
    if Digit == 0 :
        print(out, end = " ")
        return
    rem = Digit % 10
    Digit = Digit // 10
    text = getString(rem)
    for S in text :
        PrintKeypad(Digit, S + out)

PrintKeypad(79)
```

```python
def SubsetsSumkHelper(arr, Si, k) :
    if Si == len(arr) :
        if k==0 :
            return [list()]
        else :
            return list()
    Sout1 = SubsetsSumkHelper(arr, Si+1, k)
    Sout2 = SubsetsSumkHelper(arr, Si+1, k-arr[Si])
    out = (len(Sout1)+len(Sout2)) * [0]
    Index = 0
    for i in range(len(Sout1)) :
        out[Index] = Sout1[i]
        Index+=1
    for i in range(len(Sout2)) :
        out[Index] = (len(Sout2[i])+1) * [0]
        out[Index][0] = arr[Si]
        for j in range(len(Sout2[i])) :
            out[Index][j+1] = Sout2[i][j]
        Index+=1
    return out   
    
def printListOfList(liOfLi) :
    for li in liOfLi :
        for elem in li :
            print(elem, end = " ")
        print()

def SubsetSumk(arr, k) :
    return SubsetsSumkHelper(arr, 0, k)

liOfLi = SubsetSumk([3,2,5,1,4,6], 8)
printListOfList(liOfLi)     
```

```python
def subsetSumtok(arr,k):
  n=len(arr)
  if n == 1:
    if arr[0]==k:
      output = [ [k] ]
      return output
    else:
      output = []
      return output
  
  output1 = subsetSumtok(arr[:-1], k)
  output2 = subsetSumtok(arr[:-1], k - arr[-1])

  for ele in output2:
    ele.append(arr[-1])

  if arr[-1] == k:
    output2.append([k])

  return output1 + output2

def printListOfList(liOfLi) :
    for li in liOfLi :
        for elem in li :
            print(elem, end = " ")
        print()

liOfLi = subsetSumtok([3,2,5,1,4,6], 8)
printListOfList(liOfLi)
```

```python
def returnSubsets(arr) :
    n = len(arr)
    if n <= 0 :
        Out= [[]]
        return Out

    Out = returnSubsets(arr[:n-1])
    OutL = len(Out)
    for i in range(OutL) :
        Out.append(Out[i].copy())
        Out[OutL+i].append(arr[n-1])

    return Out

arr = [int(i) for i in input().split()]

out = returnSubsets(arr)
print(out)
```

```python
def permutation(str):
    if len(str) == 0:
        return 
    
    output = []
    for i in range(len(str)):
        
        start = str[i]
        
        remains = str[:i] + str[i+1:]        
        permutation(remains)
        output.append(start + remains)
        
        next1 = remains[::-1]
        permutation(next1)
        output.append(start + next1 )
        
    return output

print(permutation("abc"))
```

```python
def PrintSubsets(arr, out=[]) :
    n = len(arr)
    if n == 0 :
        for ele in out :
            print(ele, end=" ")
        print()
        return
    
    PrintSubsets(arr[1:], out)
    NewOut = out.copy()
    NewOut.append(arr[0])
    PrintSubsets(arr[1:], NewOut)
    return

n = int(input())
arr = [int(i) for i in input().split()]
PrintSubsets(arr) 
```

```python
def PrintSubsetSumtoK(arr, k, out=[]) :
    n = len(arr)
    if n == 0 :
        return
    if n == 1 :
        if arr[0] == k :
            print(k, *out)
            return
    PrintSubsetSumtoK(arr[:-1], k, out)
    PrintSubsetSumtoK(arr[:-1], k-arr[-1], [arr[-1]] + out )
    if arr[-1] == k :
        print(k, *out)

arr = [int(i) for i in input().split()]
k = int(input())
PrintSubsetSumtoK(arr, k)
```

```python
def PrintPermutatins(Str, out="") :
    if len(Str) == 0 :
        print(out)
        return
    for i in range(len(Str)) :
        PrintPermutatins(Str[:i]+Str[i+1:], out + Str[i])
Str = input()
PrintPermutatins(Str)
```
