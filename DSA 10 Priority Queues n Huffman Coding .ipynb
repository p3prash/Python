```python
class PriorityQueueNode:
    def __init__(self,ele,priority):
        self.ele = ele
        self.priority = priority
        
class PriorityQueue:
    def __init__(self):
        self.pq = []
    
    def isEmpty(self):
        return self.getSize() == 0
    
    def getSize(self):
        return len(self.pq)

    def getMin(self):
        if self.isEmpty():
            return None
        return self.pq[0].ele
    
    def __percolateUp(self):
        childIndex = self.getSize() - 1
        while childIndex > 0:
            parentIndex = (childIndex-1)//2
            
            if self.pq[parentIndex].priority > self.pq[childIndex].priority:
                self.pq[parentIndex],self.pq[childIndex] = self.pq[childIndex],self.pq[parentIndex]
                childIndex = parentIndex
            else:
                break
    
    def __percolateDown(self) :
        PI = 0
        lCI = PI*2 + 1
        rCI = PI*2 + 2
        while lCI < self.getSize() :
            MinI = PI
            if self.pq[MinI].priority > self.pq[lCI].priority :
                MinI = lCI
            if rCI < self.getSize() and self.pq[MinI].priority > self.pq[rCI].priority :
                MinI = rCI
            if MinI != PI :
                self.pq[MinI], self.pq[PI] = self.pq[PI], self.pq[MinI]
                PI = MinI
                lCI = PI*2 + 1
                rCI = PI*2 + 2
            else :
                break
                
    def insert(self,ele,priority):
        pqNode = PriorityQueueNode(ele,priority)
        self.pq.append(pqNode)
        self.__percolateUp()
        
    def removeMin(self):
        out = self.pq[0].ele
        self.pq[0] = self.pq[self.getSize()-1]
        self.pq.pop()
        self.__percolateDown()
        return out
```

```python
def downHeapify(arr, i, n) :
    PI = i 
    lCI = 2*PI+1 
    rCI = 2*PI+2
    while lCI < n :
        MinI = PI
        if arr[MinI] > arr[lCI] :
            MinI = lCI
        if rCI < n and arr[MinI] > arr[rCI] :
            MinI = rCI
        if MinI == PI :
            return
        arr[MinI], arr[PI] = arr[PI], arr[MinI]
        PI = MinI
        lCI = 2*PI + 1
        rCI = 2*PI + 2
    
def heapSort(arr) :
    n = len(arr)
    for i in range(n//2-1,-1,-1) :
        downHeapify(arr, i, n)
        
    for j in range(n-1, 0, -1) :
        arr[j], arr[0] = arr[0], arr[j]
        downHeapify(arr, 0, j)

n = input()
arr = [int(ele) for ele in input().split()]
heapSort(arr)
for ele in arr :
    print(ele,end=' ')
```

```python
import heapq, os

class BinaryTreeNode :
    def __init__(self, value, freq) :
        self.value = value
        self.freq = freq
    
    def __lt__(self, other) :
        return self.freq > other.freq

    def __eq__(self, other) :
        return self.freq == other.freq

class HuffmanCoding :
    def __init__(self, path) :
        self.path = path
        self.__heap = [] 
        self.__codes = {}
        self.reverse_Codes = {}
    
    def __make_Freq_Dict(self, text) :
        F_dict = {}
        for char in text :
            if char not in F_dict :
                F_dict[char] = 0
            F_dict[char] += 1
        return F_dict
    
    def __BuildHeap(self, F_dict) :
        for Key in F_dict :
            Freq = F_dict[Key]
            BTNode = BinaryTreeNode(Key, Freq) 
            heapq.heappush(self.__heap, BTNode)

    def __BuildTree(self) :
        while len(self.__heap) > 1 :
            BTN_1 = heapq.heappop(self.__heap)
            BTN_2 = heapq.heappop(self.__heap)
            Freq_Sum = BTN_1.freq + BTN_2.freq
            newNode = BinaryTreeNode(None, Freq_Sum)
            newNode.left = BTN_1
            newNode.right = BTN_2
            heapq.heappush(self.__heap, newNode)
        return
    
    def __BuildCOdesHelper(self, root, curr_bits ) :
        if root is None :
            return
        if root.value is not None :
            self.__codes[root.value] = curr_bits
            self.reverse_Codes[curr_bits] = root.value
            return
        self.__BuildCOdesHelper(root.left, curr_bits+"0")
        self.__BuildCOdesHelper(root.right, curr_bits+"1")
        
    def __BuildCodes(self) :
        root = heapq.heappop(self.__heap)
        self.__BuildCOdesHelper(root, "")

    def __getEncodedText(self, text) :
        Encoded = ""
        for char in text :
            Encoded += self.__codes[char]
        return Encoded
    
    def __getpaddedEncodedText(self, Encoded) :
        Padded_Amt = 8 - (len(Encoded) % 8 ) 
        for i in range(Padded_Amt) :
            Encoded += '0' 
        Padded_info = "{0:08b}".format(Padded_Amt)
        Padded_Encoded = Padded_info + Encoded
        return Padded_Encoded
 
    def __getBytesArray(self, Padded_Encoded) :
        Arr = []
        for i in range(0, len(Padded_Encoded), 8) :
            byte = Padded_Encoded[i:i+8] 
            Arr.append(int(byte, 2))
        return Arr

    def compress(self) :
        # get file from path
        # read text from file
        file_Name, file_Extension = os.path.splitext(self.path) 
        output_path = file_Name + ".bin"
        with open(self.path, 'r+') as file, open(output_path, 'wb') as output :
            text = file.read()
            text = text.rstrip()

        # make freq dict using the text
        F_dict = self.__make_Freq_Dict(text) 

        # construct the heap from freq Dict
        self.__BuildHeap(F_dict)

        # construct the binary tree from heap 
        self.__BuildTree()

        # construct the codes from binary tree
        self.__BuildCodes()

        # creating the encoded text using the codes
        Encoded = self.__getEncodedText(text)

        # put this encoded text in the binary file
        padded_Encoded = self.__getpaddedEncodedText(Encoded)
        
        # return this binary file as output
        Bytes_Arr = self.__getBytesArray(padded_Encoded)
        final_Bytes = bytes(Bytes_Arr)

        output.write(final_Bytes)
        print("compressed")
        return Output_path
    
    def __remove_Padding(self, text) :
        padded_info = text[:8]
        extra_padding = int(padded_info, 2)

        text = text[8:]
        text_after_removal_extra_padding = text[:-1*extra_padding]
        return text_after_removal_extra_padding

    def __decoded_text(self, text) :
        Decoded_text = ""
        Curr_bits = ""
        for bit in text :
            Curr_bits += bit
            if Curr_bits in self.__reverseCodes :
                Char = self.__reverseCodes[Curr_bits]
                Decoded_text += Char
                Curr_bits = ""
        return Decoded_text 


    def Decompress(self, input_path) :
        file_Name, file_Extension = os.path.splitext(self.path)
        output_path = file_Name + "_decompressed" + ".txt"
        with open(input_path, 'rb') as file, open(output_path, 'w') as output :
            bit_string = ""
            byte = file.read(1)
            while byte :
                byte = ord(byte)
                bits = bin(byte)[2:].rjust(8, '0')
                bit_string += bits
                byte = file.read(1)
            actual_text = self.__remove_Padding(bit_string)
            Decompressed_text = self.__decoded_text(actual_text)
            output.write(Decompressed_text)

path = "my-paragraphs.txt"
H = HuffmanCoding(path)
Output_path = H.compress()
```
