```python
class Node :
    def __init__(self, data) :
        self.__data = data
        self.next = None

class Stack :
    def __init__(self) :
      self.__data = []

    def push(self,item) :
      self.__data.append(item)
    
    def pop(self) :
      if self.isEmpty() is True :
        return -1
      else :  
        return self.__data.pop()
    
    def top(self) :
      if len(self.__data)== 0 :
        return -1
      else :
        return self.__data[len(self.__data) - 1]

    def size(self) :
      return len(self.__data)

    def isEmpty(self) :
      if self.size() == 0 :
        return True
      else :
        return False  
```

```python
class Node :

    def __init__(self, data) :
        self.data = data
        self.next = None

class Stack : 
    def __init__(self) :
        self.head = None
        self.count = 0
        
    '''----------------- Public Functions of Stack -----------------'''
    def push(self, data):
        if self.head is None:
            self.head = Node(data)
            self.count = self.count + 1
        else:
            new_node = Node(data)
            new_node.next = self.head
            self.head = new_node
            self.count = self.count + 1 

    def pop(self) :
        #Implement the pop() function
        if self.head is None :
            return -1
        else :
            data = self.head.data  
            self.head = self.head.next 
            self.count = self.count - 1 
            return data

    def top(self) :
        #Implement the top() function
        if self.isEmpty() is True :
            return -1
        else :
            data = self.head.data
            return data
       
    def getSize(self) :
        return self.count


    def isEmpty(self) :
        if self.head == None :
            return True
        else :
            return False
```

```python

def isBalanced(string) :

  Q = []
  for char in string :
    if char in '{([' :
      Q.append(char)
    
    elif char == '}' :
      if not Q or Q[-1] != '{' :
        return False
      Q.pop()  
         
    elif char == ']' :
      if not Q or Q[-1] != '[' :
        return False
      Q.pop()

    elif char == ')' :
      if not Q or Q[-1] != '(' :
        return False
      Q.pop()

  if not Q :
    return True
  else :
    return False
        
string = input()
ans = isBalanced(string)
print(ans)
```

```python
def stockSpan(price,n):
    s = []
    spans = []
    s.append(0)
    spans.append(1)
    for i in range(1,len(price)):
        if price[i]<=price[s[-1]]:
            spans.append(1)
            s.append(i)
        else:
            while len(s)>0 and price[i] > price[s[-1]]:
                s.pop()
            if len(s)==0:
                spans.append(i+1)
            else:
                spans.append(i-s[-1])
            s.append(i)
    return spans

n = int(input())
price = [int(ele) for ele in input().split()]
spans = stockSpan(price,n)
for ele in spans:
    print(ele,end= ' ')
```

```python
class Queue : 
    def __init__(self) :
        self.arr = []
        self.count = 0
        self.F = 0

    '''----------------- Public Functions of Stack -----------------'''
    def enqueue(self, data):
      self.arr.append(data)
      self.count+=1

    def dequeue(self) :
        #Implement the pop() function
        if self.isEmpty() is True :
          return -1
        else :
          ele = self.arr[self.F]
          self.F+=1
          self.count-=1
          return ele  

    def front(self) :
        if self.count == 0 :
            return -1
        else :
            return self.arr[self.F]
       
    def Size(self) :
        return self.count

    def isEmpty(self) :
        if self.count == 0 :
            return True
        else :
            return False
```

```python
## Queue using two stacks

class QueueUsing2Stacks :
  def __init__(self) :
    self.s1 = []
    self.s2 = []
  
  def enqueue(self, data) :
    while len(self.s1) != 0 :
      self.s2.append(self.s1.pop())
    self.s1.append(data)
    while len(self.s2)!=0 :
      self.s1.append(self.s2.pop())
    return

  def dequeue(self) :
    if len(self.s1)==0 :
      return -1
    return self.s1.pop()
  
  def front(self):
    if len(self.s1)==0 :
      return -1
    return self.s1[-1]

  def size(self) :
    return len(self.s1)

  def isEmpty(self) :
    return len(self.s1) == 0 
```

```python
from queue import Queue

class Stack :

    #Define data members and __init__()
    def __init__(self) :
        self.Q1 = Queue()
        self.Q2 = Queue()
        self.curr_size = 0
 
    
    def push(self, data) :
        #Implement the push(element) function
        self.Q1.put(data)
        self.curr_size += 1

    def pop(self) :
        #Implement the pop() function
        #popped = None
        if self.Q1.empty() is True :
            return -1
        while self.Q1.qsize() > 1 :
            self.Q2.put(self.Q1.get())
        popped = self.Q1.get()
        self.curr_size -= 1
        temp = self.Q1
        self.Q1 = self.Q2
        self.Q2 = temp
        return popped


    def top(self) :
        #Implement the top() function
        if self.Q1.empty() is True :
            return -1
        while self.Q1.qsize() > 1 :
            self.Q2.put(self.Q1.get())
        top = self.Q1.get()
        
        temp = self.Q1
        
        self.Q1 = self.Q2
        self.Q2 = temp
        self.Q1.put(top)
        return top

    
    def getSize(self) :
        #Implement the getSize() function
        return self.curr_size


    def isEmpty(self) :
        #Implement the isEmpty() function
        if self.curr_size == 0 :
            return True
        else :
            return False

```
