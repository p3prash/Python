```python
data0 = np.genfromtxt('training_diabetes_x_y_train.csv',delimiter=',')
data1 = np.genfromtxt('test_diabetes_x_test.csv',delimiter=',')
```

```python
X_train = data0[:,0:10] # we used to make our input to be 2d as expected by the LinearRegression model
Y_train = data0[:,-1]
X_test = data1[:,0:10]
Y_test = data1[:,-1]
```

```python
Y_train.shape
```

```python
 from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
diabetes_y_pred = model.predict(X_test)
```

```python
np.savetxt('diabetes Y Pred.csv',diabetes_y_pred,fmt='%.5f')
```
