```python
def Selection_Sort(arr) :
    l = len(arr)
    for i in range(l) :
        min = i
        for j in range(i,l) :
            if arr[min] > arr[j]:
                min = j

        arr[min],arr[i] = arr[i],arr[min]
    return arr  
arr = list(map(int,input().strip().split()))
print(Selection_Sort(arr))        
```

```python
def Insertion_Sort(arr):
    l = len(arr)
    for i in range(1,l) :
        j = i-1                   # 0,1,2,3,.....(l-1)
        temp = arr[i]
        while (j>=0 and arr[j]>temp) :
            arr[j+1] = arr[j]
            j-=1
        arr[j+1] = temp
    return arr
arr = list(map(int,input().strip().split()))
print(Insertion_Sort(arr))   
```

```python
def Merge(arr1,arr2,arr) :
    i,j,k = 0,0,0
    while i<len(arr1) and j<len(arr2) :
        if arr1[i] < arr2[j] :
            arr[k] = arr1[i]
            i+=1; k+=1
        else:
            arr[k] = arr2[j]
            j+=1; k+=1
  
    while i<len(arr1) :
        arr[k] = arr1[i]
        i+=1; k+=1
    while j<len(arr2) :
        arr[k] = arr2[j]
        j+=1; k+=1

def Merge_Sort(arr) :
    l = len(arr)
    if l==0 or l==1 :
        return
    mid = l//2
    arr1 = arr[:mid]
    arr2 = arr[mid:]

    Merge_Sort(arr1)
    Merge_Sort(arr2)

    Merge(arr1,arr2,arr)
    return arr

arr = list(map(int,input().strip().split()))
print(Merge_Sort(arr))   
```

```python
def Bubble_Sort(arr) :
  l = len(arr)
  for i in range(l) :
    for j in range(l-1) :
      if arr[j] > arr[j+1] :
        arr[j],arr[j+1] = arr[j+1],arr[j]
  return arr 
arr = list(map(int,input().strip().split()))
print(Bubble_Sort(arr))         
```

```python
import time
def Power(x,n) :
  if n==0 :
    return 1
   
  ans = Power(x,n//2)
  if n%2==0 :
    return ans * ans
  else:
    return ans * ans * x  
    
x = int(input())
n = int(input())
S = time.time()   
print(Power(x,n))  
E = time.time()
print(E-S) 
```

```python
def BinarySearch(arr,x,Si,Ei):
  if Si>Ei :
    return -1

  mid = (Si+Ei)//2
  if arr[mid] == x :
    return mid
  elif arr[mid] > x :
    return BinarySearch(arr,x,Si,mid-1)
  else:
    return BinarySearch(arr,x,mid+1,Ei)
arr = list(map(int,input().strip().split(", "))) 
print(BinarySearch(arr, x=12, Si=0, Ei=len(arr)-1))
```

```python
def Array_Equilibrium_Index(arr) :
  for i in range(len(arr)) :
    s1, s2 = 0, 0

    for j in range(i) :
      s1+=arr[j]
    for j in range(i+1,len(arr)) :
      s2+=arr[j]

    if s1 == s2 :
      return i
    
  return -1
arr = [int(i) for i in input().strip().split()]
print(Array_Equilibrium_Index(arr))
```

```python
def Array_Equi_Index(arr) :
  
  total_sum = 0
  for i in range(len(arr)) :
    total_sum+=arr[i]
  
  L_sum, index = 0, 0
  while index<len(arr) :
    R_sum = total_sum - L_sum - arr[index]
    if L_sum == R_sum :
      return index
    L_sum+=arr[index]
    index+=1

  return -1
arr = [int(i) for i in input().strip().split()]
print(Array_Equi_Index(arr))
```

```python
def Ouick_Sort(arr, Si, Ei) :
  if len(arr)==0 or len(arr)==1 or Si>=Ei :
    return
  
  c=0
  pivot = arr[Si]
  for i in range(Si, Ei+1) :
    if arr[i] < pivot :
      c+=1   
  arr[Si],arr[Si+c] = arr[Si+c],arr[Si]

  i,j = Si,Ei
  while i<j :
    if arr[i] < pivot :
      i+=1
    elif arr[j] >= pivot :
      j-=1
    else:
      arr[i],arr[j] = arr[j],arr[i]
      i+=1
      j-=1
          
  Ouick_Sort(arr, Si, Si+c-1)
  Ouick_Sort(arr, Si+c+1, Ei)

arr=list(int(i) for i in input().strip().split(' '))
Ouick_Sort(arr, 0, len(arr)-1)
print(*arr)
```

```python
def pairSum(arr, n, num) :
	#Your code goes here
    x = arr.sort()
    count = 0
    i,j = 0,len(arr)-1
    while i<j :
        
        if x[i]+x[j] == num :
            if x[i]==x[i+1] and x[j]==x[j-1] :
                for j in range(i+1,n) :
                    count+=1; i+=1 
            else:   
                if x[i]!=x[i+1] and x[j]!=x[j-1] :
                    i+=1; j-=1 ; count+=1
                elif x[i]==x[i+1] and x[j]!=x[j-1] :
                    i+=1 ; count+=1
                elif x[i]!=x[i+1] and x[j]==x[j-1] :
                    j-=1 ; count+=1 
               
        elif x[i]+x[j] < num :
            i+=1
        else:
            j-=1

    return count  
    
arr = [int(i) for i in input().strip().split()]
print(pairSum(arr, len(arr), num=10))
```

```python
def find3Numbers(A, n, sum): 
  
    # Sort the elements  
    A.sort() 
    count = 0
    for i in range(0, n-2): 
        l = i + 1 
          
        # index of the last element 
        r = n-1 
        while (l < r): 
          if( A[i] + A[l] + A[r] == sum): 
            count+=1
            l+=1
            r-=1
              
          elif (A[i] + A[l] + A[r] < sum): 
            l += 1
          else: # A[i] + A[l] + A[r] > sum 
            r -= 1
  
    return count
A = [int(i) for i in input().strip().split()]
sum = int(input())
print(find3Numbers(A, len(A), sum))    
```

```python
def pairSum(arr,s,e,pair_sum):
    pair_count=0
    while s<e:
        if(arr[s]+arr[e]<pair_sum):
            s=s+1
        elif(arr[s]+arr[e]>pair_sum):
            e=e-1
        else:
            if(arr[s]==arr[e]):
                total_ele=(e-s)+1
                pair_count=pair_count+(total_ele*(total_ele-1)//2)          
                return pair_count
            
            temp_s=s+1
            temp_e=e-1
            while((temp_s<=temp_e) and (arr[temp_s] == arr[s])):
                temp_s+=1
                
            while((temp_e>=temp_s) and (arr[temp_e] == arr[e])):
                temp_e-=1
                
            total_start=temp_s-s
            total_end=e-temp_e
            pair_count=pair_count+(total_start*total_end)
            s=temp_s
            e=temp_e
            
    return pair_count

def tripletSum(arr,n,sum):
    arr.sort()
    trip_count=0
    for i in range(n):
        P_sum=sum-arr[i]
        P_count=pairSum(arr,i+1,n-1,P_sum)
        trip_count+=P_count
        
    return trip_count
 
arr = [int(i) for i in input().strip().split()]
sum = int(input())   
print(tripletSum(arr, len(arr), sum))  
```

```python
def swapElements (arr,start,end):
    arr[start],arr[end]=arr[end],arr[start]
    
def reverse(arr,start,end):
    while (start<end):
        swapElements(arr,start,end)
        start +=1
        end -= 1
        
def rotate (arr,n,d):
    if n ==0:
        return
    
    if d>=n and n!=0:
        d = d%n
        
    reverse(arr,0,n-1)
    reverse(arr,0,n-d-1)
    reverse(arr,n-d,n-1)

def printList(arr, n) : 
    for i in range(n) :
        print(arr[i], end = " ")
    print()

arr = [int(i) for i in input().strip().split()]    
rotate(arr, len(arr), 5)
printList(arr, len(arr))
```
