```python
def isPalindrome(string) :

    a = 0
    b = -1
    while (a+b != len(string)):
      for i in range(0,len(string)):
          if string[a] != string[b]:
            return False
          else:
              a = a + 1
              b = b - 1
      return True
                 
string = input()
if isPalindrome(string) :
    print('true')
else :
    print('false')
```

```python
def printSubstrings(string) :
	# Your ciode goes here
    Lis = []
    L = len(string)
    for i in range(L) :
        for j in range(i,L) :
            Lis.append(string[i:j+1])
    return Lis
    
string = input()
Lis = printSubstrings(string)
for i in Lis :
    print(i, end=' ')
```

```python
def swapAlternate(arr, n) :
    #Your code goes here
    for i in range(0,n-1,2):
        temp = arr[i]
        arr[i] = arr[i+1]
        arr[i+1] = temp
    return arr

n = int(input())
arr = [int(i) for i in input().split()]
arr = swapAlternate(arr, n)
for i in arr :
  print(i,end=' ')
```

```python
def arrayEquilibriumIndex(arr, n) :
    Total = sum(arr)
    LS = 0
    for i, num in enumerate(arr) :
        Total -= num
        if LS == Total :
            return i
        LS+=num
        
    return -1

arr = [int(i) for i in input().split()]
print(arrayEquilibriumIndex(arr, len(arr)))
```

```python
def rowWiseSum(mat, nRows, mCols):
    if nRows == 0 :
        return 
    for i in range(nRows) :
        sum = 0
        for j in range(mCols) :
            sum = sum + mat[i][j]
        print(sum, end=' ')

n = int(input())
mat = [[int(i) for i in input().split()] for j in range(n)]
rowWiseSum(mat, len(mat), len(mat[0]))
```

```python
def wavePrint(mat, nRows, mCols) :
    for i in range(mCols) :
        if i%2==0 :
            for j in range(nRows) :
                print(mat[j][i], end=' ')
        
        else :
            for j in range(nRows-1,-1,-1) :
                print(mat[j][i],end=' ')

n = int(input())
mat = [[int(i) for i in input().split()] for j in range(n)]
wavePrint(mat, len(mat), len(mat[0]))
```

```python
def removeConsecutiveDuplicates(string):
    n = len(string)
    if n == 0 :
        return string
    
    out = ""
    Si = 0
    while Si < n :
        char = string[Si] 
        I = Si + 1
        
        while I < n and (char == string[I]) :
            I += 1
        out += char
        Si = I
    return out

string = input()
print(removeConsecutiveDuplicates(string))
```

```python
def RemoveConDups(S) :
    n = len(S)
    if n==0 :
        return S
    out = ""
    out += S[0]
    for i in range(n-1) :
        if S[i] == S[i+1] :
            continue
        else :
            out += S[i+1]
    return out
S = input()
print(RemoveConDups(S))
```

```python
def ReverseWord(string, S, E) :
    reverse = "" 
    while S<E :
        reverse = string[S] + reverse
        S+=1
    return reverse

def ReverseEachWord(string) :
    n = len(string)
    PSI = -1
    ans = ""
    i = 0
    while i<n :
        if string[i] == ' ' :
            ans += (ReverseWord(string, PSI + 1, i) + " ")
            PSI=i
        i+=1
    ans += (ReverseWord(string, PSI + 1, i) + " ")
    return ans

string = input()
print(ReverseEachWord(string))
```

```python
def highestOccuringChar(string):
	count = [0] * 256 
	max = -1
	out = ""

	for i in string: 
		count[ord(i)]+=1 
	for i in string: 
		if max < count[ord(i)]: 
			max = count[ord(i)] 
			out = i 
	return out

string = input()
print(highestOccuringChar(string))
```

```python
def arrange(arr, n) :
    #Your code goes here
    val=1
    start =1
    end =-1
    while(val<=n):
        if val%2!=0:
            arr[start-1]=val
            start+=1
        else:
            arr[end]=val
            end-=1
        val+=1
    return arr

def printList(arr, n) :
    for i in range(n) :
        print(arr[i], end = ' ')
    print()

n = int(input())
arr = n * [0]
arrange(arr, n)
printList(arr, n)
```

```python
def MaxSum(X, Y):
    i,j = 0, 0
    m, n = len(X), len(Y)
    maxsum, S1, S2 = 0, 0, 0

    while (i < m and j < n):
        if (X[i] < Y[j]):
            S1 = S1 + X[i]
            i+=1
            
        elif (X[i] > Y[j]):
            S2 = S2 + Y[j]
            j+=1
            
        else:
            maxsum += max(S1,S2) + X[i]
            S1=0; S2=0
            i+=1; j+=1

    while i < m:
        S1 += X[i]
        i += 1

    while j < n:
        S2 += Y[j]
        j += 1
 
    maxsum += max(S1, S2)
    return maxsum     

m = int(input())
X = list(map(int,input().strip().split()))[:m]
n = int(input())
Y = list(map(int,input().strip().split()))[:n]
print(MaxSum(X, Y))
```

```python
def findLargest(arr, nRows, mCols) :
    Rmax = -2147483648
    Rindex = 0
    for i in range(nRows) :
        Rsum = 0
        for j in range(mCols) :
            Rsum += arr[i][j]
        if Rsum > Rmax :
            Rmax = Rsum
            Rindex = i
    
    Cmax = -2147483648
    Cindex = 0
    for j in range(mCols) :
        Csum = 0
        for i in range(nRows) :
            Csum += arr[i][j]
        if Csum > Cmax :
            Cmax = Csum
            Cindex = j
    
    if Rmax >= Cmax :
        print("row", str(Rindex), str(Rmax))
    else :
        print("column", str(Cindex), str(Cmax))

n = int(input())
mat = [[int(i) for i in input().split()] for j in range(n)]
findLargest(mat, len(mat), len(mat[0]))
```

```python
def spiralPrint(mat, nRows, mCols) :
    rs, cs = 0, 0
    while rs<nRows and cs<mCols :
        for j in range(cs, mCols) :
            print(mat[rs][j], end = ' ')
        rs+=1
        
        for i in range(rs, nRows) :
            print(mat[i][mCols-1], end=' ')
        mCols-=1
        
        if rs < nRows :
            for j in range(mCols-1, cs-1, -1) :
                print(mat[nRows-1][j], end=' ')
            nRows-=1
        
        if cs < mCols :
            for i in range(nRows-1, rs-1, -1) :
                print(mat[i][cs], end=' ')
            cs+=1
n = int(input())
mat = [[int(i) for i in input().split()] for j in range(n)]
spiralPrint(mat, len(mat), len(mat[0]))
```
