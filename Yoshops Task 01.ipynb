```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
from urllib.request import urlopen as uReq

%matplotlib inline
```

```python
# yoshop's homepage
pages = 11
data = []
for i in range(pages+1) :
    yoshop_url = "https://www.yoshops.com/products?page="+str(i)
    uClient = uReq(yoshop_url)
    yoshop_html = uClient.read()
    uClient.close()
    
# Parse the whole HTML page using BeautifulSoup
    yoshop_soup = BeautifulSoup(yoshop_html, 'html.parser')
    data += yoshop_soup.select(".product")
    
cleen_data = []  # data after processing will be saved here
for i in data:
    link = "https://www.yoshops.com/"+i.select_one(".product-link")["href"]
    title = i.select_one(".product-title").get_text()
    temp = i.select_one(".product-price").get_text().split("₹")
    old_price = temp[1]
    new_price = temp[2]
    div = i.find('div', {"class": "product-thumb-inner"})
    image_url = div.find('img').attrs['src']
    has_image = "true" if not("noimage" in image_url) else "false"
    cleen_data.append([link, title, old_price, new_price, has_image])

df = pd.DataFrame(cleen_data)
df.columns = ["link", "title", "old_price", "new_price", "has_image"]
df.to_excel("yoshops task 01.xlsx")
```
