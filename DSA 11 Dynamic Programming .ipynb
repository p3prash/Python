```python
# Recursive Fibonacci using Dynamic programming
def Fibo(n, Dp) :
    if n==0 or n==1 :
        return n
    if Dp[n-1] == -1 :
        ans1 = Fibo(n-1, Dp)
        Dp[n-1] = ans1
    else :
        ans1 = Dp[n-1]
    if Dp[n-2] == -1 :
        ans2 = Fibo(n-2, Dp)
        Dp[n-2] = ans2
    else :
        ans2 = Dp[n-2]
    
    return ans1 + ans2

n = int(input())
Dp = [-1 for i in range(n+1)]
print(Fibo(n, Dp))
```

```python
# iterative fibonacci using Dynamic Programming
def FiboI(n) :
    DP = [0 for i in range(n+1)]
    DP[0] = 0
    DP[1] = 1
    i = 2
    while i <= n :
        DP[i] = DP[i-1] + DP[i-2]
        i+=1
    return DP[n]
    
n = int(input())
print(FiboI(n))
  
```

```python
# Minimum Steps to 1 using iterative DP
import sys
def MinimumStepsto1(n) :
    if n == 1 :
        return 0

    dp = [0] * (n+1)
    dp[1] = 0
    for i in range(2, (n + 1)) :
        ans1 = sys.maxsize
        ans2 = sys.maxsize
        ans3 = sys.maxsize
        
        ans3 = dp[i-1]
        if i%2 == 0 :
            ans2 = dp[i // 2]
        if i%3 == 0 :
            ans1 = dp[i // 3]
        dp[i] = 1 + min(ans1, ans2, ans3)

    return dp[n]

n = int(input())
print(MinimumStepsto1(n))
```

```python
import sys
def minSquares(n, dp) :
    if n==0 :
        return 0
    dp[0] = 0
    i = 1
    ans = sys.maxsize
    while i**2 <= n :
        if dp[n-(i**2)] == -1 :
            Value = minSquares(n-(i**2), dp)
            currAns = 1 + Value
            dp[n-(i**2)] = Value
       
        else :
            currAns = 1 + dp[n-(i**2)]
        i+=1
        ans = min(ans, currAns)
    return ans

n = int(input())
dp = [-1 for i in range(n+1)]
print(minSquares(n, dp))
```

```python
import sys, math
def MinimumSqueres(n) :
    DP = [-1] * (n+1)
    DP[0] = 0
    
    for i in range(1, n+1) :
        ans = sys.maxsize
        root = int(math.sqrt(i))
        for j in range(1, root+1) :
            CurrAns = 1 + DP[i-(j**2)]
            ans = min(CurrAns, ans)
        DP[i] = ans
    return DP[n]

n = int(input())
print(MinimumSqueres(n))
```

```python
def MaxLIS(Arr, x, n, dp) :
    if x == n :
        return 0, 0
    including_Max = 1
    for j in range(x+1, n) :
        if Arr[j] >= Arr[x] :
            if dp[j] == -1 :
                ans = MaxLIS(Arr, j, n, dp)
                dp[j] = ans
                Further_including_max = ans[0]
            else :
                Further_including_max = dp[j][0]

            including_Max = max(including_Max, 1+Further_including_max)
    if dp[x+1] == -1 :
        ans = MaxLIS(Arr, x+1, n, dp)
        dp[x+1] = ans
        excluding_Max = ans[1]
    else :
        excluding_Max = dp[x+1][1]

    overall_max = max(including_Max, excluding_Max)
    return including_Max, overall_max 

n = int(input())
Arr = [int(i) for i in input().split()]
dp = [-1 for i in range(n+1)]
print(MaxLIS(Arr, 0, n, dp)[1])
```

```python
def LISIter(arr) :
    n = len(arr)
    dp = [[0 for j in range(2)] for i in range(n+1)]
    for i in range(n-1, -1, -1) :
        including_Max = 1
        for j in range(i+1, n) :
            if arr[j] >= arr[i] :
                including_Max = max(including_Max, 1+dp[j][0])
        dp[i][0] = including_Max
        excluding_Max = dp[i+1][1]
        overall_max = max(including_Max, excluding_Max)
        dp[i][1] = overall_max
    
    return dp[0][1]

n = int(input())
arr = [int(i) for i in input().split()]
print(LISIter(arr))
```

```python
import sys
def LCSrec(str1, str2, dp, i=0, j=0) :
    
    if i == len(str1) or j == len(str2) :
        return 0

    if str1[i] == str2[j] :
        if dp[i][j] == -1 :
            out = LCSrec(str1, str2, dp, i+1, j+1)
            dp[i][j] = out
            ans = 1 + out
        else :
            ans = 1 + dp[i][j]
    
    else :
        if  dp[i+1][j] == -1 :
            ans1 = LCSrec(str1, str2, dp, i+1, j)
            dp[i+1][j] = ans1
        else :
            ans1 = dp[i+1][j]

        if dp[i][j+1] == -1 :
            ans2 = LCSrec(str1, str2, dp, i, j+1)
            dp[i][j+1] = ans2
        else :
            ans2 = dp[i][j+1]

        ans = max(ans1, ans2)
    return ans

str1 = input()
str2 = input()
dp = [[-1 for i in range(len(str2)+1)] for j in range(len(str1)+1)]
print(LCSrec(str1, str2, dp))
```

```python
import sys
def LCSIterative(str1, str2) :
    m, n = len(str1), len(str2)
    dp = [[0 for i in range(n+1)] for j in range(m+1)]
    for i in range(m-1, -1, -1) :
        for j in range(n-1, -1, -1) :
            if str1[i] == str2[j] :
                dp[i][j] = 1 + dp[i+1][j+1]
            else :
                ans1 = dp[i+1][j]
                ans2 = dp[i][j+1]
                dp[i][j] = max(ans1, ans2)
        
    return dp[0][0]

str1 = input()
str2 = input()
print(LCSIterative(str1, str2))
```

```python
def KnapsackR(W, Wt, val, i=0) :
    n = len(Wt)
    if i==n :
        return 0
    if Wt[i] > W :
        ans = KnapsackR(W, Wt, val, i+1)
    else :
        ans1 = KnapsackR(W, Wt, val, i+1)
        ans2 = val[i] + KnapsackR(W-Wt[i], Wt, val, i+1)
        ans = max(ans1, ans2)
    return ans

Wt = [int(i) for i in input().split()]
val = [int(j) for j in input().split()]
W = int(input())
print(KnapsackR(W, Wt, val))
```

```python
def KnapsackIter(W, Wt, Val) :
    n = len(Val)
    dp = [[0 for j in range(W+1)] for i in range(n+1)]
    for i in range(1, n+1) :
        for j in range(1, W+1) :
            if j < Wt[i-1] :
                ans = dp[i-1][j]
            else :
                ans1 = val[i-1] + dp[i-1][j-Wt[i-1]]
                ans2 = dp[i-1][j]
                ans = max(ans1, ans2)
            dp[i][j] = ans
    return dp[n][W]

Wt = [int(i) for i in input().split()]
val = [int(j) for j in input().split()]
W = int(input())
print(KnapsackIter(W, Wt, val))
```

```python
import sys
def MatChainMRecur(arr, i, j, dp) :
    if i==j :
        return 0
    min_Val = sys.maxsize
    for k in range(i, j) :
        if dp[i][k] == -1 :
            ans1 = MatChainMRecur(arr, i, k, dp)
            dp[i][k] = ans1
        else :
            ans1 = dp[i][k]
        
        if dp[k+1][j] == -1 :
            ans2 = MatChainMRecur(arr, k+1, j, dp)
            dp[k+1][j] = ans2
        else :
            ans2 = dp[k+1][j]

        mCost = arr[i-1] * arr[k] * arr[j]
        Curr_Val = ans1 + ans2 + mCost
        min_Val = min(min_Val, Curr_Val)
    return min_Val

arr = [int(i) for i in input().split()]
n = len(arr) - 1
dp = [[-1 for i in range(n+1)] for j in range(n+1)]
print(MatChainMRecur(arr, 1, n, dp))
```
