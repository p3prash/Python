```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
sns.set()
```

```python
Boston = datasets.load_boston()
x = Boston.data
y = Boston.target
print(x.shape, y.shape)
```

```python
df = pd.DataFrame(x)
print(Boston.feature_names)
df.columns = Boston.feature_names
df.describe()
```

```python
df["CRIM_Sq"] = df.CRIM ** 2
df["ZN_Sq"] = df.ZN ** 2
df["INDUS_Sq"] = df.INDUS ** 2
df["CHAS_Sq"] = df.CHAS ** 2
df["NOX_Sq"] = df.NOX ** 2
df["RM_Sq"] = df.RM ** 2
df["DIS_Sq"] = df.DIS ** 2
df["RAD_Sq"] = df.RAD ** 2
df["TAX_Sq"] = df.TAX ** 2
df["PTRATIO_Sq"] = df.PTRATIO ** 2
df["LSTAT_Sq"] = df.LSTAT ** 2
X1 = df.values
print(X1.shape)
df.describe()
```

```python
from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y)
x1_train, x1_test, y1_train, y1_test = model_selection.train_test_split(X1, y)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
print(x1_train.shape, x1_test.shape, y1_train.shape, y1_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
Algo1 = LinearRegression()
Algo2 = LinearRegression()
```

```python
Algo1.fit(x_train, y_train)
Algo2.fit(x1_train, y1_train)
```

```python
y_pred = Algo1.predict(x_test)
y2_pred = Algo1.predict(x_test)
Train_S = Algo1.score(x_train, y_train)
Test_S = Algo1.score(x_test, y_test)
print("Train Score : ", Train_S)
print("Test Score : ", Test_S)
```

```python
y2_pred = Algo2.predict(x1_test)
Train_S1 = Algo2.score(x1_train, y1_train)
Test_S1 = Algo2.score(x1_test, y1_test)
print("Train Score 2 : ", Train_S1)
print("Test Score 2 : ", Test_S1)
```
