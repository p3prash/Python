```python
def NAme_Correction(x) :
    if x == '':
        x = 'CUSTOMER' 
    return x

def Price(value) :
    if value == '0.00' :
        value = '0.01'
    return value

def State(y) :
    if len(y) == 2 :
        if y.isupper() == True :
            return y
        else :
            return y.upper()

```

```python
new['Billing First Name'] = new['Billing First Name'].apply(NAme_Correction)
new['Billing Last Name'] = new['Billing Last Name'].apply(NAme_Correction)
new['Shipping First Name'] = new['Shipping First Name'].apply(NAme_Correction)
new['Shipping Last Name'] = new['Shipping Last Name'].apply(NAme_Correction)
new['Shipping State or Province'] = new['Shipping State or Province'].apply(State)
new['Billing State or Province'] = new['Billing State or Province'].apply(State)
new['Order Total'] = new['Order Total'].apply(Price)
new['Unit Price'] = new['Unit Price'].apply(Price)
new
```
