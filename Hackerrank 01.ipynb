```python
def plusMinus(arr):
    n = len(arr)
    p,z,N = 0,0,0 
    for i in range(n) :
        if arr[i] > 0 :
            p+=1
        elif arr[i] == 0 :
            z+=1
        else :
            N+=1
    print('%.5f'%(p/n))
    print('%.5f'%(N/n))
    print('%.5f'%(z/n))
arr = [int(i) for i in input().split()]
plusMinus(arr)
```

```python
def flippingMatrix(matrix):
    n = len(matrix)
    s = 0
    for i in range(n//2):
        for j in range(n//2):

            s += max(matrix[i][j], matrix[i][n-j-1], matrix[n-i-1][j], matrix[n-i-1][n-j-1])
    return s

if __name__ == '__main__':
    q = int(input().strip())
    for q_itr in range(q):
        n = int(input().strip())

        matrix = []
        for x in range(2*n):
            matrix.append(list(map(int, input().rstrip().split())))
        print(flippingMatrix(matrix))
```

```python
def findZigZagSequence(a, n):
    a.sort()
    mid = int((n+1)/2)-1
    a[mid], a[n-1] = a[n-1], a[mid]

    st = mid + 1
    ed = n - 2
    while(st <= ed):
        a[st], a[ed] = a[ed], a[st]
        st = st + 1
        ed = ed - 1

    for i in range (n):
        if i == n-1:
            print(a[i])
        else:
            print(a[i], end = ' ')
    return

test_cases = int(input())
for cs in range (test_cases):
    n = int(input())
    a = list(map(int, input().split()))
    findZigZagSequence(a, n)
```

```python
def encrypt(text,s):
    result = ""
    for i in range(len(text)):
        char = text[i]
        if (char.isupper()):
            result += chr((ord(char) + s - 65) % 26 + 65)

        else:
            result += chr((ord(char) + s - 97) % 26 + 97)
    return result
text = input()
s = int(input())
print(encrypt(text,s))
```

```python
def caesarCipher(s, k):
    result = ""
    for i in range(len(s)) :
        char = s[i]
        if char.isalpha() is False  :
            result += char
        elif (char.isupper()) :
            result += chr((ord(char)+k-65)%26 + 65)
        else :
            result += chr((ord(char)+k-97)%26 + 97)
    
    return result

if __name__ == '__main__':
    n = int(input().strip())
    s = input()
    k = int(input().strip())
    result = caesarCipher(s, k)
    print(result)
```

```python
class VendingMachine:
    def __init__(self, num_items, item_price) :
        self.num_items = num_items
        self.item_price = item_price
        
    def buy(self, req_items, money) :
            if self.num_items > req_items and (self.item_price*req_items <= money ) :
                self.num_items -= req_items
                return (money - self.item_price*req_items )
        
            if self.num_items < req_items :
                print("Not enough items in the machine")
                return self.num_items
            
            if (money < self.item_price*req_items) :
                print("Not enough coins")
                return (money - self.item_price*req_items )
```

```python
def transformSentence(s):
    words = s.split()
    result = ""
    for i in words :
        result += i[0]
        for j in range(1, len(i)) :
            if i[j-1].lower() < i[j].lower() :
                result += i[j].upper()
            elif i[j-1].lower() > i[j].lower() :
                result += i[j].lower()
            else :
                result += i[j]
        result += " "
    return result[:-1:]
    
s = input()
print(transformSentence(s))
```

```python
def palindromeIndex(s):
    l = len(s)
    i = 0
    j = l-1
    while i<l:
        if s[i]!=s[j]:
            break
        i+=1
        j-=1
    if i>j: return -1
    a = i+1
    b = j
    while a<j and b>i+1:
        if s[a]!=s[b]:
            return j
        a+=1
        b-=1
    return i

for i in range(int(input())):
    print(palindromeIndex(input()))
```

```python
def super_Digit(x):
    if len(x) <= 1:
        return x
    else:
        return super_Digit(digitSum(x))

def digitSum(x):
    return str(sum((int(i) for i in list(x))))

def super_Digits(x,k):
    a = digitSum(x)
    return super_Digit(str(int(a)*k))

n, k = input().split()
print( super_Digits(n, int(k)))
```

```python
def minimum_bribes(q):
    q = [i-1 for i in q]  # set queue to start at 0
    bribes = 0
    
    for i, o in enumerate(q):
        cur = i 
        if o - cur > 2:
            print("Too chaotic")
            return
        for k in q[max(o - 1, 0):i]:
            if k > o:
                bribes += 1

    print(bribes)
q = [int(i) for i in input().split()]
minimum_bribes(q)
```
