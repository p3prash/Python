```python
import sqlite3
db = sqlite3.connect('Employee.sqlite')
```

```python
cur = db.cursor()
```

```python
cur.execute('create table EmployeeDetail (employeeid int Primary Key, name Text, age int, department Text, salary int);')
```

```python
cur.execute('insert into EmployeeDetail values (100, "Rishav", 21, "Marketing", 50000 );')
cur.execute('insert into EmployeeDetail values (101, "Abhishek", 23, "Operations", 60000 );')
cur.execute('insert into EmployeeDetail  values (102, "Aman", 25, "Human Resources", 75000 );')
```

```python
db.commit()
```

```python
Cursor = cur.execute('select * from EmployeeDetail;')
for row in Cursor :
    print(row)
```
