```python
def printPathHelper(x, y, Maze, n, Out) :
    if x==n-1 and y==n-1 :
        Out[x][y]=1
        print(Out)
        return 
    
    if x<0 or y<0 or x>=n or y>=n or Maze[x][y] == 0 or Out[x][y] == 1 :
        return
        
    Out[x][y] = 1
    printPathHelper(x+1, y, Maze, n, Out)
    printPathHelper(x, y+1, Maze, n, Out)
    printPathHelper(x-1, y, Maze, n, Out)
    printPathHelper(x, y-1, Maze, n, Out)
    Out[x][y] = 0
    return

def printPath(Maze) :
    n = len(Maze)
    Out = [[0 for j in range(n)] for i in range(n)]
    printPathHelper(0, 0, Maze, n, Out)

n = int(input())
Maze = []
for i in range(n) :
    row = [int(ele) for ele in input().split()]
    Maze.append(row)
printPath(Maze)
```

```python
def isSafe(row, col, n, Board) :
    i = row -1
    while i >= 0 :
        if Board[i][col] == 1 :
            return False
        i-=1
    
    i = row-1 ; j = col-1
    while i>=0 and j>=0 :
        if Board[i][j] == 1 :
            return False
        i-=1 ; j-=1
    
    i = row-1 ; j = col+1
    while i>=0 and j<n :
        if Board[i][j] == 1 :
            return False
        i-=1 ; j+=1
    
    return True

def PrintNqueensHelper(row, n, Board ) :
    if row == n :
        for i in range(n) :
            for j in range(n) :
                print(Board[i][j], end=" ")
        print()
        return
    for col in range(n) :
        if isSafe(row, col, n, Board) :
            Board[row][col] = 1
            PrintNqueensHelper(row+1, n, Board )
            Board[row][col] = 0
    return
            
def PrintNqueens(n) :
    Board = [[0 for j in range(n)] for i in range(n)]
    PrintNqueensHelper(0, n, Board)

n = int(input())
PrintNqueens(n)
```

```python
N = 9
def isSafe(grid, row, col, num):

    for x in range(9):
        if grid[row][x] == num:
            return False

    for x in range(9):
        if grid[x][col] == num:
            return False

    startRow = row - row % 3
    startCol = col - col % 3
    for i in range(3):
        for j in range(3):
            if grid[i + startRow][j + startCol] == num:
                return False
    return True

def solveSudoku(grid, row, col):

    if (row == N - 1 and col == N):
        return True
    
    if col == N:
        row += 1
        col = 0

    if grid[row][col] > 0:
        return solveSudoku(grid, row, col + 1)
    for num in range(1, N + 1, 1):

        if isSafe(grid, row, col, num):

            grid[row][col] = num
 
            if solveSudoku(grid, row, col + 1):
                return True

        grid[row][col] = 0
    return False


grid = [[ int(ele) for ele in input().split() ]for i in range(9)]
ans = solveSudoku(grid,0,0)
if ans is True:
    print('true')
else:
    print('false')
```
