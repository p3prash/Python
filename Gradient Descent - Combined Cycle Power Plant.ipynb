```python
training = np.loadtxt('training_ccpp_x_y_train.csv', delimiter=',')
testing = np.loadtxt('test_ccpp_x_test.csv', delimiter=',')
```

```python
def Step_Gradient(X, Y, m, lr):
    m_slope = np.zeros(len(X[0]))
    for i in range(len(X)):
        x=X[i]
        y=Y[i]
        for j in range(len(x)):
            m_slope[j] += (-2/len(X))*(y-sum(m*x))*x[j]
    new_m = m - (lr * m_slope)
    return new_m
```

```python
def Cost_Function(m, x, y):
    Cost = 0
    for i in range(len(x)):
        Cost += (1/len(x)) * ((y[i]-sum(m*x[i]))**2)
    print(Cost)
```

```python
def GD(x, y, lr, Iter) :
    m=np.zeros(len(x[0]))
    for i in range(Iter) :
        m=Step_Gradient(x, y, m, lr)
        print("itr= ", i, "cost=", end=' ')
        Cost_Function(m, x, y)
    return m
```

```python
def Gradient_Descent(x, y, Iter, lr) :
    x = np.append(x, np.ones(len(x)).reshape(-1, 1), axis=1)
    m=GD(x, y, lr, Iter)
    return m
```

```python
x=training[:, :-1]
y=training[:, -1]

#adding squared values of each column

sq=[]
for i in range(len(x[0])):
    for j in range(i, len(x[0])):
        sq.append(x[:, i]*x[:, j])
sq=np.array(sq)
for i in sq:
    x=np.append(x, i.reshape(-1, 1), axis=1)
    
#scaling

scaler=preprocessing.StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
m = Gradient_Descent(x, y, Iter=1000, lr=0.115)
```

```python
sq=[]
for i in range(len(testing[0])):
    for j in range(i, len(testing[0])):
            sq.append(testing[:, i]*testing[:, j])
sq=np.array(sq)
for i in sq:
    testing=np.append(testing, i.reshape(-1, 1), axis=1)
```

```python
testing1=scaler.transform(testing)
x_test = np.append(testing1, np.ones(len(testing1)).reshape(-1, 1), axis=1)
```

```python
ans=[]
for i in x_test:
    ans.append(sum(i*m))
for i in ans:
    print(i)
ans=np.array(ans)
np.savetxt(X=ans,fname='Data_Power_Plant.csv' , delimiter=',', fmt='%.5f')
```
