```python
import csv
with open("year2017.csv") as csvFile :
    data = csv.DictReader(csvFile, skipinitialspace = True)
    listData = list(data) 
country_killed = {}
for row in listData :
    key = row['Country']
    value = row['Killed']
    if value != '' :
        value = int(float(value))
    else :
        value = 0
    
    if key in country_killed :
        country_killed[key] += value
    else :
        country_killed[key] = value
print(country_killed, end='')
```

```python
import csv
with open('amazon_jobs_dataset.csv') as csvFile :
    data = csv.DictReader(csvFile, skipinitialspace = True )
    listData = list(data)

monthD = []
for row in listData :
    a = row['Posting_date'].split()
    if '2018' in a :
        monthD.append(a)

dict = {}
for i in monthD:
    if i[0] in dict.keys():
        dict[i[0]]+=1
    else:
        dict[i[0]]=1

freq=0
maxmonth='January'
for i in dict :
    if dict[i]>freq:
        freq=dict[i]
        maxmonth=i
print(maxmonth, freq) 
```

```python
with open("amazon_jobs_dataset.csv") as csvFile :
    data = csv.DictReader(csvFile, skipinitialspace=True )
    listData = list(data)

count = 0
Dict = {}
Dict['Java'] = 0
Dict['C++'] = 0
Dict['Python'] = 0
for row in listData :
    Loc = row['location'].strip().split(',')
    x = row['BASIC QUALIFICATIONS']
    if 'Java' in x and ('Bachelor' in x or 'BS' in x or 'BA' in x) and (Loc[0] == 'IN') :
        Dict['Java'] += 1
    elif 'C++' in x and ("Bachelor" in x or "BA" in x or "BS" in x) and (Loc[0]=='IN') :
        Dict['C++'] += 1
    elif 'Python' in x and ("Bachelor" in x or "BA" in x or "BS" in x) and (Loc[0]=='IN') :
        Dict['Python'] += 1

count=0
language=""
for lan in Dict.keys():
    if count<Dict[lan]:
        count=Dict[lan]
        language=lan
        
print(language, count)
```

```python
import csv
with open("amazon_jobs_dataset.csv") as csvFile :
    data = csv.DictReader(csvFile, skipinitialspace=True )
    listData = list(data)
    dict={}
    for row in listData :
        a=row['BASIC QUALIFICATIONS']
        b=row['location'].split(",")[0]
        if "Java" in a:
            if b in dict.keys():
                dict[b]+=1
            else:
                dict[b]=1
                
count=0
country=""
for i in dict:
    if dict[i]>count:
        count=dict[i]
        country=i
print(country, count)
```
