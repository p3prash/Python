```python
import numpy as np
input_=np.arange(1,21,1)
input_=input_.reshape(4,5)

p = input_[2, 0:3]
for i in p:
  print(i,end=' ')
print('\n')  
q = input_[1:4, 3]
for i in q:
  print(i,end=' ')
print('\n')   
r = input_[2:4]
for i in r:
    for j in i:
      print(j,end=' ')
print('\n')   
s = input_[1:3, 1:3]
for i in s:
  for j in i:
      print(j,end=' ')
```

```python
import numpy as np
def Identity(m,n):
    for row in range(0, m):
        for col in range(0, n):
 
            # Here end is used to stay in same line
            if (row == col):
                print("1 ", end=" ")
            else:
                print("0 ", end=" ")
        print()

Identity(5,6)    
```
