```python
def insert_at_Ith(head, i, data) :
  if i < 0 or i > length(head) :
    return head
  
  count = 0
  prev = None
  curr = head
  while count < i :
    prev = curr
    curr = curr.next
    count+=1

  new_Node = Node(data)
  if prev is not None :
    prev.next = new_Node
  else :
    head = new_Node

  new_Node.next = curr

  return head

head = take_Input()
head = insert_at_Ith(head, 0, 69)
printLL(head)
```

```python
# appending last n nodes to first

def appendLastNToFirst(head, n):
    L = length(head)
     
    if (L != n and n < L):
        skip = L - n
        prev = None
        curr = head
         
        # Skip the nodes
        while (skip > 0):
            prev = curr
            curr = curr.next
            skip-=1
             
        # Change the pointers
        prev.next = None
        tempHead = head
        head = curr
         
        # Find the last node
        while (curr.next != None):
            curr = curr.next
        curr.next = tempHead

    return head
head = take_Input()
out = appendLastNToFirst(head, 3)    
printLL(out)  
```

```python
def Merge_LL(head1,head2) :
  F_head, F_tail = None, None
  if head1 is None and head2 is None :
    return head1
  if head1 is not None and head2 is None :
    return head1  
  if head1 is None and head2 is not None :
    return head2

  if head1.data < head2.data :
    F_head, F_tail = head1, head1
    head1 = head1.next
  else :
    F_head, F_tail = head2, head2   
    head2 = head2.next

  while head1 is not None and head2 is not None :
    if head1.data < head2.data :
      F_tail.next = head1
      F_tail = F_tail.next
      head1 = head1.next 
    else :
      F_tail.next = head2
      F_tail = F_tail.next
      head2 = head2.next

  if head1 is not None :
    F_tail.next = head1
  else :
    F_tail.next = head2

  return F_head

head1 = take_Input()
head2 = take_Input()
head = Merge_LL(head1,head2)
printLL(head)
```

```python
def mergeLists(headA, headB):
    dummyNode = Node(0)

    tail = dummyNode
    while True:

        if headA is None:
            tail.next = headB
            break
        if headB is None:
            tail.next = headA
            break
        if headA.data <= headB.data:
            tail.next = headA
            headA = headA.next
        else:
            tail.next = headB
            headB = headB.next

        tail = tail.next

    return dummyNode.next
headA = take_Input()
headB = take_Input()
head = mergeLists(headA, headB)
printLL(head)    
```

```python
def Midpoint_LL(head) :
  if head is None :
    return head
  slow, fast = head, head
  while fast.next != None and fast.next.next != None :
    slow = slow.next
    fast = fast.next.next
  return slow

def MergeSort_LL(head) :
  if head.next is None :
    return head
  head1 = head
  mid = Midpoint_LL(head)
  head2 = mid.next
  mid.next = None
  head1 = MergeSort_LL(head1)
  head2 = MergeSort_LL(head2)
  
  head = Merge_LL(head1, head2)
  return head
head = take_Input()
head = MergeSort_LL(head)
printLL(head)  
```

```python
def skipMdeleteN(head, M, N):
    curr = head

    while(curr):
        # Skip M nodes
        for count in range(1, M):
            if curr is None:
                return
            curr = curr.next
                  
        if curr is None :
            return

        skip = curr.next
        for count in range(1, N+1):
            if skip is None:
                break
            skip = skip.next

        curr.next = skip
        curr = skip
    return head    

head = take_Input()
head = skipMdeleteN(head, M=2, N=3)
printLL(head)
```

```python
def kReverse(head, k) :
	#Your code goes here
    if k==0 :
        return head
    
    if head == None:
       return None
    curr = head
    prev = None
    count = 0

    # Reverse first k nodes of the linked list
    while(curr is not None and count < k):
        next = curr.next
        curr.next = prev
        prev = curr
        curr = next
        count += 1
    head.next = kReverse(next, k)

    return prev

head = take_Input()
head = kReverse(head, k=4)
printLL(head)
```
