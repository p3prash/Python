```python
def MaxSubarray(arr,n) :
     
    max_so_far = arr[0]
    max_ending_here = 0
     
    for i in range(n):
        max_ending_here = max_ending_here + arr[i]
        if max_ending_here < 0:
            max_ending_here = 0
         
        elif (max_so_far < max_ending_here):
            max_so_far = max_ending_here
             
    return max_so_far
    
n = int(input())
arr = [int(i) for i in input().split()]
print(MaxSubarray(arr,n))
```
