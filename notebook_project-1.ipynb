```python
data=pd.read_csv("training_boston_x_y_train.csv")
training_dataset=data.copy()
training_dataset=training_dataset.drop(" Y",axis=1)
training_dataset["area_square"]=training_dataset[" ZN"]**2
training_dataset["room_square"]=training_dataset[" RM"]**2
training_dataset["crime_square"]=training_dataset["# CRIM"]**2
training_dataset["tax_square"]=training_dataset[" TAX"]**2
training_dataset["AGE_sq"]=training_dataset[" AGE"]**2

training_dataset["C"]=1

training_dataset["Y"]=data[" Y"]
```

```python
def step_gd(data,learning_rate,m):
    
    m_slope=np.zeros(data.shape[1]-1)
    
    N=len(data)
    
    for i in range(N):
        x=data[i,:-1]
        y=data[i,-1]
        
        m_slope+=(-2/N)*((y-(m*x).sum())*x)
        
    m=m-(learning_rate*m_slope)
    
    return m
```

```python
def costfun(data,m):
    
    M = len(data)
    
    total_cost=0
    
    for i in range(M):
        
        x=data[i,:-1]
        
        y=data[i,-1]
        
        total_cost+=(1/M)*((y-(m*x).sum())**2)
        
    return total_cost
```

```python
def predict(m,x,c):
    result=[]
    N=len(x)
    for i in range(N):
        result.append((m*x[i]).sum()+c)
        
    return result
```
