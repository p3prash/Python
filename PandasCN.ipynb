```python
def largetRowSumAndIndex(arr):
    n = len(arr)
    m = len(arr[0])
    
    max_s = -2147483648
    max_i = 0
    i = 0
    while (i < n):
        sum = 0
        for j in range(m -1):
            sum += arr[i][j]
            i +=1
        if sum > max_s:
            max_i = i
            max_s = sum
            
        elif sum == max_s:
            if max_i > i:
                max_i = i
                max_s = sum       
    return max_i,max_s

def largetColSumAndIndex(arr):
    n = len(arr)
    m = len(arr[0])
    
    max_s = -2147483648
    max_i = 0
    j = 0
    while (j < m ):
        sum = 0
        for i in range(n-1):
            sum += arr[i][j]
            j += 1
        if sum > max_s:
            max_i = j
            max_s = sum       
    return max_i, max_s

def findlargest(l,m,n):
    arr = [ [ l[(j*n)+i] for i in range(n)] for j in range(m)]
    row = largetRowSumAndIndex(arr)
    col = largetColSumAndIndex(arr)

    if row[1] > col[1]:
        print('row',end =' ')
        print(row[0],end =' ')
        print(row[1],end =' ')    
    elif col[1] > row[1]:
        print('column',end =' ')
        print(col[0],end =' ')
        print(col[1],end =' ')    
    else:
        print('row',end =' ')
        print(row[0],end =' ')
        print(row[1],end =' ')

m, n=(int(i) for i in input().strip().split(' '))
l=[int(i) for i in input().strip().split(' ')]
findlargest(l,m,n)
```
