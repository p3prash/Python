```python
# Set Matrix Zeroes
class Solution(object):
   def setZeroes(self, Mat):
      n, m = len(Mat),len(Mat[0])
      flag = False
      if Mat[0][0] == 0:
         flag = True; row = False; Col = False
      for i in range(1,n):
         if Mat[i][0] == 0:
            Col = True
            break
      for j in range(1,m):
         if Mat[0][j] == 0:
            row = True
            break
      for i in range(1,n):
         for j in range(1,m):
            if Mat[i][j] == 0:
               Mat[0][j] = 0
               Mat[i][0]=0
      for i in range(1,n):
         for j in range(1,m):
            if not Mat[i][0] or not Mat[0][j]:
               Mat[i][j] = 0
      if flag:
         for i in range(n):
            Mat[i][0] = 0
         for i in range(m):
            Mat[0][i]=0
      else:
         if Col:
            for i in range(n):
               Mat[i][0]=0
         if row:
            for i in range(m):
               Mat[0][i]=0
      return Mat
ob1 = Solution()
print(ob1.setZeroes([[1,0,1,1],[1,1,1,1],[0,1,1,1]]))
```
