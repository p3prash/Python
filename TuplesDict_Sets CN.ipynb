```python
def most_frequent(List):
    counter = 0
    num = List[0]
     
    for i in List:
        curr_frequency = List.count(i)
        if(curr_frequency> counter):
            counter = curr_frequency
            num = i
 
    return num

n = int(input())
List = list(map(int,input().strip().split()))[:n]
print(most_frequent(List))
```

```python
 # Function to find maximum distance between equal elements
def maxDist(arr, n):
    d = {}
    
    m = 0
    for i in range(n):
        if arr[i] not in d :
            d[arr[i]] = i
 
        else:
            m = max(m, i-d[arr[i]])
 
    return m
 
if __name__=='__main__':
    n = int(input())
    arr = list(map(int,input().split()))
    
    print(maxDist(arr, n))
```

```python
# sum for numbers in a set by considering only once

def SumUnique(l):
  s = set()
  for i in l:
    s.add(i)
  sum = 0
  for i in s:
    sum = sum + i
  return sum 

SumUnique([4,5,6,5,5,4,8,4,1,2,3,4,5,5,6,7,9,8,4,5])    
```

```python
# printing non repeated character in a string

from collections import Counter
def printNonrepeated(string):
  freq = Counter(string)
# Traverse the string
  for i in string:
    if(freq[i] == 1):
      print(i)
      break
 
string =input()
printNonrepeated(string)
```

```python
# print the number which is occuring even no of times using time of o(n**2)

def evenCount(arr):
    n = len(arr)
    for i in range(n):
        count = 0
        for j in range(n):
            if arr[j] == arr[i]:
                count += 1
        if (count%2 == 0):
            return arr[i]
            
    return -1
# Main
n=int(input())
arr=list(int(i) for i in input().strip().split(' '))
print(evenCount(arr))
```

```python
# print the number which is occuring even no of times using hashing
def evenCounter(arr):
    n = len(arr)
    H = dict()
  
    for i in range(n):
        H[arr[i]]=H.get(arr[i],0) + 1
    
    for i in H:
        if(H[i]% 2 == 0):
            return i
    return -1
# Main
n=int(input())
arr=list(int(i) for i in input().strip().split(' '))
print(evenCounter(arr))    
```

```python
def uniqueTriangle(arr):
  row = len(arr)
  col = len(arr[0])
  mp = {}
  hel = {}

  for i in range(row):
    tri = arr[i]
    tri.sort()
    strA = [str(x) for x in tri]
    strB = ''
    strB = strB.join(strA)
    if strB not in mp.values():
      mp[i] = strB
    else:
      hel[i] = strB

  count = 0
  for i in range(row):
    if i in mp:
      val = mp.get(i)
      if val not in hel.values():
        count = count + 1

  print (count)
n = int(input())
arr = [list(map(int, input().split())) for x in range(n)]
uniqueTriangle(arr)
```

```python
def printPairDiffK(l, k):
    paircount = 0
    m = {}
    for num in l :
        if num+k in l :
            paircount += m[num+k]
        if k!=0 and num-k in l :
            paircount += m[num-k]
        
        if num in m :
            m[num] += 1
        else :
            m[num] = 1
                
    return paircount           
    
# Main
n=int(input())
l=list(int(i) for i in input().strip().split(' '))
k=int(input())
print(printPairDiffK(l, k))
```
