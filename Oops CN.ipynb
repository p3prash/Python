```python
class Person:
    def SetValues(self,name,age):
        self.name = name
        self.age = age
    
    def GetValues(self):
        print(("The name of the person is " + self.name),end=' ')
        print("and the age is " + str(self.age)+".")
        

#Driver code goes here.
name = input()
age = int(input())
s = Person()
s.SetValues(name,age)
s.GetValues()
```

```python
# Area of a Rectangle
class Rectangle:

    def __init__(self,length,breadth):
        self.length = length
        self.breadth = breadth
        
    def getArea(self):
        Area = self.length * self.breadth
        print(Area)

l,b = input().split(' ')
length = int(l)
breadth = int(b)
A = Rectangle(length,breadth)
A.getArea()
```

```python
'''
    Time complexity: O(2^N)
    Space complexity: O(N) 

    Where 'N' reprents the "Nth" number .
'''

def fibonacciNumber(n):
    
    mod = int(1e9 + 7)

    # Base Case
    if (n <= 1):
        return n
    
    # Recursive call
    return (fibonacciNumber(n - 1) + fibonacciNumber(n - 2)) % mod
```
