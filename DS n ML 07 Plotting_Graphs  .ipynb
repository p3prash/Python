```python
# plot job year
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = pd.read_csv('amazon_jobs_dataset.csv')
date = np.array(df['Posting_date'])
year = []
for item in date:
    year.append(item[-4:])
year = np.array(year)
numOfPosting = {}
for item in year:
    if numOfPosting.get(item, 0) == 0:
        numOfPosting[item] = 1
    else:
        numOfPosting[item] += 1
        
for key, value in sorted(numOfPosting.items()):
    print("{} {}".format(key, value))

lists = sorted(numOfPosting.items())
x, y = zip(*lists)
plt.plot(x, y)
plt.show()
```

```python
# plot month job
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = pd.read_csv('amazon_jobs_dataset.csv',encoding='utf-8')
date = df["Posting_date"] 
date = date.str.split(' ', expand = True)
month = date[0]
freq = month.value_counts()
x = freq.index
height = freq
plt.bar(range(len(x)), height, tick_label = x)
plt.xticks(rotation="vertical")
plt.show()
for i in range(len(x)) :
    print(x[i], height[i])
```

```python
# plot city job
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data=pd.read_csv('amazon_jobs_dataset.csv',encoding='utf-8')
location=data['location'].str.split(', ',expand=True)
arr=location[0]=='IN'
india_city=location[arr][2]
freq=india_city.value_counts()
plt.axis("equal")
x=freq.index
plt.pie(freq,labels=x,autopct='%.2f%%')
plt.show()
perc=np.true_divide(freq,freq.sum(axis=0))*100
for i in range(len(x)):
    print(x[i],format(perc[i],'.2f'))
```

```python
# plot job java
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = pd.read_csv('amazon_jobs_dataset.csv',encoding='utf-8')
year = df["Posting_date"].str.split(', ', expand = True)[1]
Basic_Q = df["BASIC QUALIFICATIONS"]
Arr = np.array(list(zip(year, Basic_Q)))
for i in range(len(Arr)) :
    if 'Java' in Arr[i][1] or 'java' in Arr[i][1] :
        Arr[i][1] = 1
    else :
        Arr[i][1] = 0
Arr = Arr[Arr[:, 1]=='1']
Arr = Arr[:, 0]
year = np.unique(Arr)
job = []
for i in year :
    job.append(len(Arr[Arr==i]))
plt.scatter(year, job, edgecolors='black', s = 100)
plt.grid()
plt.show()
for i in range(len(year)) :
    print(year[i], job[i])
```
