```python
class TreeNode() :
    def __init__(self,data) :
        self.data = data
        self.children = list()
```

```python
def PrintTree(root) :
    if root == None : # this is just an edge case 
        return
    print(root.data, end = " ")
    for child in root.children :
        PrintTree(child)
```

```python
def PrintTreeDetailed(root) :
    if root == None : # this is just an edge case 
        return
    print(root.data,end = ":")
    for child in root.children :
        print(child.data, end = ',')
    print()
    for child in root.children :
        PrintTreeDetailed(child)
```

```python
def takeTreeInput() :
    print("enter rootData:",end=' ')
    rootData = int(input())
    if rootData == -1 :
        return None
    root = TreeNode(rootData) 
    print("enter No of child of ", root.data,end=':')
    Childcount = int(input())
    for i in range(Childcount) :
        child = takeTreeInput()   
        root.children.append(child)
    return root 
```

```python
def NumNodes(root) :
    if root == None :
        return 0
    count = 1
    for child in root.children :
        count += NumNodes(child)
    return count
```

```python
def createLevelWiseTree(arr):
    root = TreeNode(int(arr[0]))
    q = [root]
    size = len(arr)
    i = 1
    while i<size:
        parent = q.pop(0)
        childCount = int(arr[i])
        i += 1
        for j in range(0,childCount):
            temp = TreeNode(int(arr[i+j]))
            parent.children.append(temp)
            q.append(temp)
        i += childCount
    return root
```

```python
def HeightofTree(root) :
    if root == None :
        return 0
    if len(root.children)==0 :
        return 1
    Height = 0
    for child in root.children :
        CHeight = HeightofTree(child)
        if CHeight > Height :
            Height = CHeight
        
    return 1+Height

arr = list(int(x) for x in input().strip().split())
root = createLevelWiseTree(arr)
print(HeightofTree(root))
```

```python
def printLevelWiseTree(tree) :
    if (tree == None):
        return
    q = []  # Create a queue
    q.append(tree); # Enqueue root
    while (len(q) != 0): 
        n = len(q)
        while (n > 0) :
            root = q[0]
            q.pop(0) 
            print(root.data, end=':')
            for child in root.children :
                q.append(child)
                if child == root.children[-1] :
                    print(child.data, end="")
                    break
                print(child.data, end=",")
            print()
            n -= 1

arr = list(int(x) for x in input().strip().split()) 
tree = createLevelWiseTree(arr)
printLevelWiseTree(tree)
```

```python
def containsX(tree, x) :
    if tree == None :
        return False
    if tree.data == x :
        return True
    for child in tree.children :
        if containsX(child, x) :
            return True
    return False

arr = list(int(x) for x in input().strip().split()) 
tree = createLevelWiseTree(arr)
x = int(input())
print(containsX(tree, x))
```

```python
def leafNodeCount(tree):
    if tree == None :
        return 
    count=0
    if tree != None and len(tree.children)==0 :
        return 1
    for child in tree.children :
        count += leafNodeCount(child)
    
    return count
arr = list(int(x) for x in input().strip().split()) 
tree = createLevelWiseTree(arr)
print(leafNodeCount(tree))
```

```python
def NodeSum(root) :
    if root == None :
        return 0
    sum = root.data
    for child in root.children :
        sum+=child.data
    return sum

arr = list(int(x) for x in input().strip().split()) 
root = createLevelWiseTree(arr)
print(NodeSum(root))
```

```python
def maxSumUtil(root, resNode, maxsum):
    if root == None:
        return
    currsum = root.data
    for child in root.children :
        currsum += child.data
        resNode, maxsum = maxSumUtil(child, resNode, maxsum)
    if currsum > maxsum:
        resNode = root
        maxsum = currsum
     
    return resNode, maxsum

def maxSum(root):
    resNode, maxsum = None, 0
    resNode, maxsum = maxSumUtil(root, resNode, maxsum)
    return resNode.data
arr = list(int(x) for x in input().strip().split()) 
root = createLevelWiseTree(arr)
print(maxSum(root))
```

```python
def isIdentical(tree1, tree2):
    if not tree1:
        if not tree2:
            return True
        return False
    if tree1.data!=tree2.data or len(tree1.children)!=len(tree2.children):
        return False
    for child1,child2 in zip(tree1.children,tree2.children):
        if not isIdentical(child1,child2):
            return False
    return True
arr1 = list(int(x) for x in input().strip().split(' '))
tree1 = createLevelWiseTree(arr1)
arr2 = list(int(x) for x in input().strip().split(' '))
tree2 = createLevelWiseTree(arr2)
if isIdentical(tree1, tree2):
    print('true')
else:
    print('false')
```

```python
def nextLargest(tree, n, arr = []) :
    if tree == None :
        return
    if tree.data > n :
        arr.append(tree.data)
    for child in tree.children :
        if child.data > n :
            arr.append(child.data)
        nextLargest(child, n, arr)
    if len(arr) != 0 :
        Min = min(arr)
    else :
        Min = None
    return Min

arr1 = list(int(x) for x in input().strip().split(' '))
tree = createLevelWiseTree(arr1)
n = int(input())
print(nextLargest(tree, n))
```

```python
def replacewithDepth(tree, L=0) :
    if tree == None :
        return
    tree.data = L
    for child in tree.children :
        child.data = L+1
        replacewithDepth(child, L+1)

def printLevelAtNewLine(tree):
    q = [tree]
    newq = []
    while q:
        parent = q.pop(0)
        print(parent.data, end=' ')
        for child in parent.children:
            newq.append(child)
        if len(q)==0:
            q = newq
            newq = []
            print()

arr = list(int(x) for x in input().strip().split(' '))
tree = createLevelWiseTree(arr)
replacewithDepth(tree)
PrintTree(tree)
print()
printLevelAtNewLine(tree)
```
