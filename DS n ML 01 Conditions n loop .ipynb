```python
def reverse(n):
  rev = 0
  while(n > 0):
      rev = rev * 10 + (n % 10)
      n = n // 10   
  return rev
n=int(input())
result = reverse(n)
print(result)
```

```python
# sum of all even and odd digits in a given number N
def SumofEvenOdd(num) :
    Sev = 0
    Sod = 0
    while num > 0 :
        rem = num%10
        if rem%2==0 :
            Sev+=rem
        else :
            Sod+=rem
        num = num//10
    return Sev, Sod
    
num = int(input())
E, O = SumofEvenOdd(num)
print(E, O)
```

```python
def checkPalindrome(num):

  isp = num
  rev = 0
  while (num > 0):
      rev = rev * 10 + num % 10
      num = num // 10
      
  if(isp == rev):
      print('true')
  else:
      print('false')

num = int(input())
checkPalindrome(num)
     
```

```python
def NthFibo(num) :
    a = 0
    b = 1
    if num == 0 :
        return a
    elif num == 1 :
        return b
    else :
        while num>1 :
            c=a+b
            a=b
            b=c
            num-=1
    return b

num = int(input())
print(NthFibo(num))
```

```python
#Write a program to print first x terms of the series 3N + 2 which are not multiples of 4.
def firstXT(x):
  count = 0 
  for i in range(1,1000):
    while count != x:
      num = 3*i + 2
      i = i + 1
      if num % 4 != 0:
        count = count + 1
        
        print(num, end=' ')

x = int(input())
firstXT(x)
```

```python
# to claculate sum of N numbers and factorial of N

def sumofn(N):
  if N < 1:
    return 0
  else:
    value = N + sumofn(N-1)
    return(value) 

def fact(N):
  if N <= 0:
    return 1
  else:
    val = N * fact(N-1)
    return(val)  

N = int(input())
C = int(input())
if C == 1:
  x = sumofn(N)
  print(x)

elif C == 2:
  x = fact(N)
  print(x)        

else:
  print(-1)
```

```python
def floorSqrt(N):
  if (N==0 or N==1):
    return N

  i=1
  r=1
  while (r <= N):
    i = i+1
    r = i*i

  return i-1

N = int(input())
x= floorSqrt(N)
print(x)

```

```python
#Write your totalPrime function here.
def totalPrime(S,E):
    p = 0
    for i in range(S,E+1):
        count = 0
        for j in range(1,i+1):
            if i%j == 0:
                count += 1
        if count == 2:
            p += 1
    return p  
#Taking S and E space seperated input.
S,E = map(int,input().split(' '))
print(totalPrime(S,E))
```

```python
def SumofN(n) :
    i=1
    sum=0
    while i<=n :
        sum+=i
        i+=1
    return sum

def ProductofN(n) :
    i=1
    Mul = 1
    while i<=n :
        Mul*=i
        i+=1
    return Mul

n = int(input())
C = int(input())
if C==1 :
    print(SumofN(n))
elif C==2 :
    print(ProductofN(n))
else :
    print('-1')
```

```python
def Dec2Bin(n) :
    out = 0
    pv=0
    while n>0 :
        rem = n%2
        out = out+rem*(10**pv)
        n=n//2
        pv+=1
    return out
      
n = int(input())
print(Dec2Bin(n))
```

```python
def Bin2Dec(n) :
    pv = 0
    ans = 0
    while n>0 :
        rem = n%10
        ans = ans + rem*(2**pv)
        pv+=1
        n=n//10
    return ans


n = int(input())
print(Bin2Dec(n))
```

```python
def EvenFiboSum(n) :
    if (n < 2) :
        return 0
    n1 = 0
    n2 = 2
    sum = n1 + n2

    while (n2 <= n) :

        n3 = 4 * n2 + n1

        if (n3 > n) :
            break

        n1 = n2
        n2 = n3
        sum = sum + n2
    
    return sum

n = int(input())
print(EvenFiboSum(n))
```
