```python
class BinaryTreeNode :
  def __init__(self, data) :
    self.data = data
    self.left = None
    self.right = None
```

```python
def PrintTreeDetailed(root) :
  if root == None :
    return
  print(root.data, end = ":") 
  if root.left != None :
    print(" L", root.left.data, end = ", ")
  if root.right != None :
    print("R", root.right.data, end = "")
  print()
  PrintTreeDetailed(root.left)
  PrintTreeDetailed(root.right)
```

```python
#Taking level-order input using fast I/O method
import queue
def takeInput():
    levelOrder = list(map(int, input().strip().split(" ")))
    start = 0
    length = len(levelOrder)
    root = BinaryTreeNode(levelOrder[start])
    start += 1

    q = queue.Queue()
    q.put(root)

    while not q.empty():
        currentNode = q.get()

        leftChild = levelOrder[start]
        start += 1

        if leftChild != -1:
            leftNode = BinaryTreeNode(leftChild)
            currentNode.left =leftNode
            q.put(leftNode)

        rightChild = levelOrder[start]
        start += 1

        if rightChild != -1:
            rightNode = BinaryTreeNode(rightChild)
            currentNode.right =rightNode
            q.put(rightNode)

    return root
X = takeInput()
PrintTreeDetailed(X)
```

```python
def searchInBST(root, x) :
    if root==None :
        return False
    if root.data == x :
        return True
    elif root.data > x :
        return searchInBST(root.left, x) 
    else :
        return searchInBST(root.right, x) 
     
root = takeInput()
x = int(input())
ans = searchInBST(root, x)
if ans :
  print('true')
else :
  print('false')
```

```python
def elementsInRangeK1K2(root, k1, k2):
    if root == None :
        return
    
    if root.data < k1 :
        elementsInRangeK1K2(root.right, k1, k2)
        
    elif root.data >= k1 and root.data <= k2 :
        elementsInRangeK1K2(root.left, k1, k2)
        print(root.data, end = ' ')
        elementsInRangeK1K2(root.right, k1, k2)
        
    else :
        elementsInRangeK1K2(root.left, k1, k2)
 
root = takeInput()
k1,k2 = [int(i) for i in input().split()]
elementsInRangeK1K2(root, k1, k2)
```

```python
def constructBST(lst): 
    l = len(lst)
    if not lst :
        return None
    mid = (l-1)//2
    root = BinaryTreeNode(lst[mid])
    root.left = constructBST(lst[:mid])
    root.right = constructBST(lst[mid+1:])
    return root 

def preOrder(root):
    if root==None:
        return
    print(root.data, end=' ')
    preOrder(root.left)
    preOrder(root.right)

lst=[int(i) for i in input().split()]
root=constructBST(lst)
preOrder(root)
```

```python
def MinTree(root) :
  if root == None :
    return 10000
  leftMin = MinTree(root.left)
  rightTree = MinTree(root.right)
  return min(leftMin, rightMin, root.data)

def MaxTree(root) :
  if root==None :
    return -10000
  leftMax = MaxTree(root.left)
  rightMax = MaxTree(root.right)
  return max(leftMax, rightMax, root.data)
  
def CheakBST(root) :
  if root==None :
    return True

  Lmax = MaxTree(root.left)
  Rmin = MinTree(root.right)
  if root.data > Rmin or root.data <= Lmax :
    return False

  isLeft = CheakBST(root.left)
  isRight = CheakBST(root.right)
  return isLeft and isRight

root = takeInput()
print(CheakBST(root))
```

```python
def isTreeBST(root) :
    if root==None :
        return 10000, -10000, True
    Lmin, Lmax, isLBST = isTreeBST(root.left)
    Rmin, Rmax, isRBST = isTreeBST(root.right)
    
    minimum = min(root.data, Lmin, Rmin)
    maximum = max(root.data, Lmax, Rmax)

    isBST = True
    if Lmin>= root.data or Rmin<=root.data :
      isBST = False
    if isLBST or isRBST :
      isBST = False
    return minimum, maximum, isBST
    
root = takeInput()
isTreeBST(root)
```

```python
def BinaryST(root, Min, Max) :
    if root == None :
        return True
    if root.data < Min or root.data > Max :
        return False
    isLinConstraints = BinaryST(root.left, Min, root.data -1)
    isRinConstraints = BinaryST(root.right, root.data, Max)

    return isLinConstraints and isRinConstraints

root = takeInput()
Min, Max = [int(i) for i in input().split()]
print(BinaryST(root, Min, Max))
```

```python
def FindPathinBST(root, S) :
    if root==None :
        return None
    if root.data == S :
        l = []
        l.append(root.data)
        return l
    Lout = FindPathinBST(root.left, S)
    if Lout != None :
        Lout.append(root.data)
        return Lout
    
    Rout = FindPathinBST(root.right, S)
    if Rout != None :
        Rout.append(root.data)
        return Rout
    else :
        return None

root = takeInput()
S = int(input())
FindPathinBST(root, S)
```

```python
class BST:  
    def __init__(self):
        self.root = None
        self.numNodes = 0
    
    def PrintTreeHelper(self, root) :
        if root == None :
            return
        print(root.data,end=':',sep='')
        if root.left != None :
            print('L:', root.left.data, end=',',sep='')
        if root.right != None :
            print('R:',root.right.data,end='',sep='')
            
        print()
        self.PrintTreeHelper(root.left)
        self.PrintTreeHelper(root.right)
        
    def printTree(self):
        self.PrintTreeHelper(self.root)
    
    def SearchHelper(self, root, data) :
        if root==None :
            return False
        if root.data == data :
            return True
        elif root.data > data :
            return self.SearchHelper(root.left, data) 
        else :
            return self.SearchHelper(root.right, data)
        
    def search(self, data):
        return self.SearchHelper(self.root, data)
        
    def insertHelper(self, root, data) :
        if root==None :
            Node = BinaryTreeNode(data)
            return Node
        
        if root.data >= data :
            root.left = self.insertHelper(root.left, data)
            return root
        
        else :
            root.right = self.insertHelper(root.right, data)
            return root
            
    def insert(self, data):
        self.numNodes += 1
        self.root = self.insertHelper(self.root, data)
    
    def minimum(self, root) :
        if root == None :
            return 10000
        if root.left == None :
            return root.data
        return self.minimum(root.left)
    
    
    def deleteHelper(self, root, data) :
        if root.data == None :
            return None
        
        if root.data > data :
            NewLeft = self.deleteHelper(root.left, data)
            root.left = NewLeft
            return root
        
        if root.data < data :
            NewRight = self.deleteHelper(root.right, data)
            root.right = NewRight
            return root
        
        if root.left == None and root.right == None :
            return None
        
        if root.left == None :
            return root.right
        
        if root.right == None :
            return root.left
        
        Replace = self.minimum(root.right)
        root.data = Replace
        NewRight = self.deleteHelper(root.right, Replace)
        root.right = NewRight
        return root
        
    def delete(self, data):
        self.numNodes -= 1
        self.root = self.deleteHelper(self.root, data)
        
    def count(self):
        return self.numNodes
```
