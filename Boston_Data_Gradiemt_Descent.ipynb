```python
def step_gradient(data, learning_rate, m):
    m_slope = np.zeros(data.shape[1]-1)
  
    M = len(data)
    for i in range(M):
        x = data[i, :-1]
        y = data[i, -1]
        m_slope += (-2/M)*((y - (x*m).sum())*x)
    
    m = m - (learning_rate * m_slope)
  
    return m
```

```python
def cost_function(data,m):
    
    N=len(data)
    total_cost=0
    
    for i in range(N):
        x=data[i,:-1]
        y=data[i,-1]
        
        total_cost+=(1/N)*((y-(m*x).sum())**2)
        
    return total_cost
```
