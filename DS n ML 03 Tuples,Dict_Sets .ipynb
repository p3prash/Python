```python
def maxDist(arr, n):
    d = {}
    
    m = 0
    for i in range(n):
        if arr[i] not in d :
            d[arr[i]] = i
 
        else:
            m = max(m, i-d[arr[i]])
 
    return m
 
if __name__=='__main__':
    n = int(input())
    arr = list(map(int,input().split()))
    
    print(maxDist(arr, n))
```

```python
# sum for numbers in a set by considering only once

def SumUnique(l):
  s = set()
  for i in l:
    s.add(i)
  sum = 0
  for i in s:
    sum = sum + i
  return sum 

SumUnique([4,5,6,5,5,4,8,4,1,2,3,4,5,5,6,7,9,8,4,5])    
```

```python
def duplicateCount(arr):
    n = len(arr)
    s = set()
    for i in arr :
        s.add(i)
    l = len(s)
    return (n-l)

# Main
n=int(input())
arr=list(int(i) for i in input().strip().split(' '))
print(duplicateCount(arr))
```

```python
# printing non repeated character in a string
from collections import Counter
def printNonrepeated(string):
  freq = Counter(string)
# Traverse the string
  for i in string:
    if(freq[i] == 1):
      print(i)
      break
 
string =input()
printNonrepeated(string)
```

```python
def freqMap(l) :
    d = {}
    for i in l :
        if i in d :
            d[i] += 1
        else :
            d[i] = 1
    return d
    
def differentNames(l) :
    d = freqMap(l)
    ans = {}
    for name in d :
        if d[name] > 1 :
            ans[name] = d[name]
    return ans
        
names=input().strip().split()
m=differentNames(names)
if m :
    for name in m :
        if m[name] > 1 :
            print(name, m[name])
else :
    print(-1)

```

```python
def evenCount(arr):
    n = len(arr)
    for i in range(n):
        count = 0
        for j in range(n):
            if arr[j] == arr[i]:
                count += 1
        if (count%2 == 0):
            return arr[i]
            
    return -1
# Main
n=int(input())
arr=list(int(i) for i in input().strip().split(' '))
print(evenCount(arr))
```

```python
def printPairDiffK(arr, k) :
    # Write your code here     
    d = {}
    count = 0
    for i in arr :
        if (i+k) in d :
            count += d[i+k]
            
        if k!=0 and (i-k) in d :
            count += d[i-k]
            
        if i in d :
            d[i] += 1
        else :
            d[i] = 1
            
    return count

n = int(input())
arr = [int(i) for i in input().split()]
k = int(input())
print(printPairDiffK(arr, k))   
```
